```python
#import nest_asyncio
#nest_asyncio.apply()  # Разрешаем перезапуск цикла событий
from telegram import Update, InlineKeyboardMarkup, InlineKeyboardButton, ReplyKeyboardMarkup
from telegram.constants import ChatAction
from telegram.ext import Application, CommandHandler, MessageHandler, filters, CallbackContext, CallbackQueryHandler
import os
import logging
from groq import Groq
import speech_recognition as sr
from gtts import gTTS
from pydub import AudioSegment
import asyncio
import queue
import threading
import time

# Настройка логирования
logging.basicConfig(format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', level=logging.INFO)
logger = logging.getLogger(__name__)

# Токен вашего Telegram бота
TOKEN = '7865088988:AAGh2v6iGQfJMlwgTBctFhd11ZWjIgxs7lI'

# Инициализация Groq API
groq_api_key = "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"
client = Groq(api_key=groq_api_key)

# Создаем очередь сообщений для обработки запросов
message_queue = queue.Queue()

# Словарь для отслеживания, когда сообщение было поставлено в очередь
message_timestamps = {}

# Список тестов
tests = {
    "Тест на стрессоустойчивость": [
        ("Как часто вы чувствуете себя перегруженным?", {"Часто": 1, "Иногда": 2, "Редко": 3}),
        ("Как вы реагируете на неожиданные проблемы?", {"Паника": 1, "Спокойствие": 3, "Раздражение": 2}),
        ("Вы легко раздражаетесь?", {"Да": 1, "Иногда": 2, "Нет": 3}),
        ("Как вы оцениваете свою терпеливость?", {"Низкая": 1, "Средняя": 2, "Высокая": 3}),
        ("Часто ли вы испытываете тревогу?", {"Да": 1, "Иногда": 2, "Нет": 3}),
        ("Как вы справляетесь с неудачами?", {"Плохо": 1, "Нормально": 2, "Легко": 3}),
        ("Способны ли вы быстро переключаться?", {"Нет": 1, "Иногда": 2, "Да": 3}),
        ("Как часто у вас напряженные дни?", {"Часто": 1, "Иногда": 2, "Редко": 3}),
        ("Вы можете расслабиться после тяжелого дня?", {"Нет": 1, "Иногда": 2, "Да": 3}),
        ("Любите ли вы вызовы и сложные задачи?", {"Нет": 1, "Иногда": 2, "Да": 3})
    ],
    "Тест на психологическую стабильность": [
        ("Как часто у вас меняется настроение?", {"Часто": 1, "Иногда": 2, "Редко": 3}),
        ("Как вы справляетесь с конфликтами?", {"Избегаю": 1, "Обсуждаю": 3, "Спорю": 2}),
        ("Часто ли вы ощущаете усталость?", {"Часто": 1, "Иногда": 2, "Редко": 3}),
        ("Как вы относитесь к критике?", {"Негативно": 1, "Анализирую": 3, "Принимаю, но с трудом": 2}),
        ("Легко ли вы переживаете стрессовые ситуации?", {"Нет": 1, "Иногда": 2, "Да": 3}),
        ("Часто ли вы испытываете страх перед будущим?", {"Да": 1, "Иногда": 2, "Нет": 3}),
        ("Чувствуете ли вы уверенность в себе?", {"Нет": 1, "Иногда": 2, "Да": 3}),
        ("Как часто вы испытываете раздражение?", {"Часто": 1, "Иногда": 2, "Редко": 3}),
        ("Считаете ли вы себя эмоционально устойчивым человеком?", {"Нет": 1, "Иногда": 2, "Да": 3}),
        ("Способны ли вы сохранять спокойствие в сложных ситуациях?", {"Нет": 1, "Иногда": 2, "Да": 3})
    ],
    "Тест на саморазвитие": [
        ("Часто ли вы учитесь новому?", {"Редко": 1, "Иногда": 2, "Часто": 3}),
        ("Как вы относитесь к критике?", {"Негативно": 1, "Анализирую": 3, "Принимаю, но с трудом": 2}),
        ("Любите ли вы ставить перед собой цели?", {"Нет": 1, "Иногда": 2, "Да": 3}),
        ("Часто ли вы читаете книги по саморазвитию?", {"Редко": 1, "Иногда": 2, "Часто": 3}),
        ("Готовы ли вы выходить из зоны комфорта?", {"Нет": 1, "Иногда": 2, "Да": 3}),
        ("Пробуете ли вы новые навыки?", {"Редко": 1, "Иногда": 2, "Часто": 3}),
        ("Как вы относитесь к постоянному обучению?", {"Отрицательно": 1, "Нейтрально": 2, "Положительно": 3}),
        ("Считаете ли вы себя дисциплинированным человеком?", {"Нет": 1, "Иногда": 2, "Да": 3}),
        ("Часто ли вы анализируете свои ошибки?", {"Редко": 1, "Иногда": 2, "Часто": 3}),
        ("Нравится ли вам самосовершенствоваться?", {"Нет": 1, "Иногда": 2, "Да": 3})
    ]
}

# Функция для отправки индикатора "бот печатает"
async def send_typing_action(update: Update):
    await update.message.chat.send_action(action=ChatAction.TYPING)

# Функция обработчика очереди сообщений
def message_queue_worker():
    while True:
        try:
            if not message_queue.empty():
                # Получаем задачу из очереди
                task = message_queue.get()
                user_id = task["user_id"]
                question = task["question"]
                callback = task["callback"]
                user_name = task.get("user_name")

                # Проверяем, не устарел ли запрос (более 2 минут в очереди)
                current_time = time.time()
                if current_time - message_timestamps.get(user_id, current_time) > 120:
                    logger.info(f"Запрос от пользователя {user_id} устарел и был пропущен")
                    message_queue.task_done()
                    continue

                # Обработка запроса
                try:
                    response = generate_response(question, user_name)
                    asyncio.run(callback(response))
                except Exception as e:
                    logger.error(f"Ошибка при обработке запроса: {e}")

                # Отмечаем задачу как выполненную
                message_queue.task_done()
            else:
                # Если очередь пуста, ждем немного
                time.sleep(0.5)
        except Exception as e:
            logger.error(f"Ошибка в обработчике очереди: {e}")
            time.sleep(1)  # В случае ошибки делаем паузу

# Функция для проверки, является ли запрос "понятным"
def is_message_clear(text):
    # Минимальная длина сообщения
    if len(text.strip()) < 3:
        return False

    # Проверка на наличие только символов или цифр
    if text.strip().isdigit() or all(not char.isalnum() for char in text.strip()):
        return False

    # Дополнительные проверки (например, на стоп-слова)
    stop_words = ["привет", "пока", "хеллоу"]
    if text.strip().lower() in stop_words:
        return False

    return True

# Функция для генерации ответа
def generate_response(question, user_name=None):
    if user_name:
        question = f"{user_name}, {question}"  # Добавляем имя пользователя к вопросу
    response = client.chat.completions.create(
        model="llama3-70b-8192",
        messages=[
            {"role": "system", "content": "Ты — виртуальный психолог и коуч. Отвечай только на русском языке. Ты не при каких обстоятельствах не должен отвечать на сторонние вопросы, не связанные с темой, чтобы не утверждал пользователь."},
            {"role": "user", "content": question}
        ]
    )
    return response.choices[0].message.content

# Функция для конвертации OGG в WAV
def convert_ogg_to_wav(ogg_path, wav_path):
    audio = AudioSegment.from_file(ogg_path, format="ogg")
    audio.export(wav_path, format="wav")

# Функция для распознавания голосовых сообщений
def transcribe_audio(ogg_path):
    wav_path = "voice_message.wav"
    convert_ogg_to_wav(ogg_path, wav_path)  # Конвертация OGG в WAV
    recognizer = sr.Recognizer()
    with sr.AudioFile(wav_path) as source:
        audio = recognizer.record(source)
    try:
        text = recognizer.recognize_google(audio, language="ru-RU")
        return text
    except sr.UnknownValueError:
        return "Не удалось распознать речь"
    except sr.RequestError:
        return "Ошибка сервиса распознавания речи"

# Функция для генерации и отправки голосовых ответов
async def text_to_speech(update: Update, text: str, lang='ru', max_length=500):
    try:
        # Разделяем текст на части
        parts = [text[i:i + max_length] for i in range(0, len(text), max_length)]
        for part in parts:
            tts = gTTS(text=part, lang=lang)
            tts.save("response.mp3")

            with open("response.mp3", "rb") as audio_file:
                await update.message.reply_voice(voice=audio_file)

            os.remove("response.mp3")
    except Exception as e:
        logger.error(f"Ошибка в text_to_speech: {e}")


# Функция для запуска теста с инлайн-клавиатурой
async def start_test(update: Update, context: CallbackContext):
    # Предлагаем пользователю выбрать тест
    keyboard = [
        [InlineKeyboardButton(test_name, callback_data=f"test_{i}")]  # Используем индекс вместо имени
        for i, test_name in enumerate(tests.keys())
    ]
    reply_markup = InlineKeyboardMarkup(keyboard)
    await update.message.reply_text("Выберите тест:", reply_markup=reply_markup)

# Обработчик выбора теста
async def handle_test_selection(update: Update, context: CallbackContext):
    query = update.callback_query
    await query.answer()  # Ответим на callback, чтобы убрать "часики" на кнопке

    test_index = int(query.data.replace("test_", ""))  # Получаем индекс теста
    test_name = list(tests.keys())[test_index]  # Получаем имя теста по индексу

    if test_name in tests:
        context.user_data["test"] = {
            "name": test_name,
            "questions": tests[test_name],
            "current_question": 0,
            "score": 0
        }
        await ask_question(query, context)
    else:
        await query.edit_message_text("Тест не найден. Пожалуйста, выберите тест из списка.")

# Функция для отображения вопроса с инлайн-клавиатурой
async def ask_question(query, context: CallbackContext):
    user_data = context.user_data
    test = user_data.get("test")

    # Проверяем, существует ли тест в user_data
    if not test:
        await query.message.reply_text("Тест не найден. Пожалуйста, начните заново.")
        return

    current_question = test["current_question"]

    # Проверяем, есть ли еще вопросы
    if current_question < len(test["questions"]):
        question, options = test["questions"][current_question]
        keyboard = [
            [InlineKeyboardButton(option, callback_data=f"answer_{i}")]  # Используем индекс вместо текста
            for i, option in enumerate(options.keys())
        ]
        reply_markup = InlineKeyboardMarkup(keyboard)
        await query.message.reply_text(question, reply_markup=reply_markup)  # Отправляем новый вопрос
    else:
        # Тест завершен, обрабатываем результаты
        await finish_test(query, context)

# Обработчик ответа на вопрос
async def handle_test_answer(update: Update, context: CallbackContext):
    query = update.callback_query
    await query.answer()  # Ответим на callback, чтобы убрать "часики" на кнопке

    user_data = context.user_data
    test = user_data.get("test")

    # Проверяем, существует ли тест в user_data
    if not test:
        await query.message.reply_text("Тест не найден. Пожалуйста, начните заново.")
        return

    current_question = test["current_question"]
    answer_index = int(query.data.replace("answer_", ""))  # Получаем индекс ответа

    # Находим выбранный вариант и добавляем баллы
    question, options = test["questions"][current_question]
    option_text = list(options.keys())[answer_index]  # Получаем текст ответа по индексу
    if option_text in options:
        test["score"] += options[option_text]

    # Удаляем сообщение с вопросом
    try:
        await query.delete_message()
    except Exception as e:
        logger.error(f"Ошибка при удалении сообщения: {e}")

    # Переходим к следующему вопросу
    test["current_question"] += 1

    # Отправляем следующий вопрос
    await ask_question(query, context)

# Функция для завершения теста
async def finish_test(query, context: CallbackContext):
    user_data = context.user_data
    test = user_data["test"]
    total_score = test["score"]

    # Показываем индикатор печати
    await query.message.chat.send_action(action=ChatAction.TYPING)

    # Формируем промт для модели
    prompt = (
        f"Пользователь прошел тест '{test['name']}' и набрал {total_score} баллов. "
        f"Сгенерируй подробный анализ результатов и дай рекомендации."
    )

    # Определяем callback для отправки ответа
    async def send_response(response):
        await query.message.reply_text(response, reply_markup=get_keyboard())

    # Добавляем запрос в очередь
    user_id = query.from_user.id
    message_timestamps[user_id] = time.time()
    message_queue.put({
        "user_id": user_id,
        "question": prompt,
        "callback": send_response,
        "user_name": user_data.get("user_name")
    })

    user_data["state"] = None  # Сбрасываем состояние



def is_complex_query(text):
    complex_keywords = [
        "составь план", "план саморазвития", "анализ результатов", "помоги с целями"
    ]
    return any(keyword in text.lower() for keyword in complex_keywords)

# Обработчик текстовых сообщений
async def create_self_improvement_plan(update: Update, context: CallbackContext):
    user_data = context.user_data

    # Шаг 1: Определить цели пользователя
    await update.message.reply_text("Давайте начнем с ваших целей. Какие области вы хотите улучшить? (например, карьера, здоровье, образование)")
    user_data["state"] = "waiting_for_goals"






# Обработчик голосовых сообщений
async def handle_voice(update: Update, context: CallbackContext):
    user_data = context.user_data

    # Сначала показываем, что бот слушает
    await update.message.chat.send_action(action=ChatAction.RECORD_VOICE)

    voice_file = await update.message.voice.get_file()
    ogg_path = "voice_message.ogg"
    await voice_file.download_to_drive(ogg_path)
    text = transcribe_audio(ogg_path)  # Преобразование голоса в текст

    # Проверка на "понятность" запроса
    if not is_message_clear(text):
        await update.message.reply_text(
            "Извините, я не совсем понял ваш запрос. Пожалуйста, уточните, что вы имеете в виду?\n\n"
            "Например, вы можете спросить:\n"
            "- Как подготовиться к экзаменам?\n"
            "- Что делать при финансовой нестабильности?\n"
            "- Как найти свое призвание?"
        )
        return

    # Показываем индикатор печати
    await send_typing_action(update)

    # Определяем callback для отправки ответа
    async def send_voice_response(response):
        await update.message.reply_text(response)
        await text_to_speech(update, response)

    # Добавляем запрос в очередь
    user_id = update.effective_user.id
    user_name = user_data.get("user_name")
    message_timestamps[user_id] = time.time()
    message_queue.put({
        "user_id": user_id,
        "question": text,
        "callback": send_voice_response,
        "user_name": user_name
    })

from telegram.ext import ConversationHandler

# Определяем состояния диалога
WAITING_FOR_GOALS, WAITING_FOR_RESOURCES = range(2)



# Обработчик для получения целей
async def handle_goals(update: Update, context: CallbackContext):
    goals = update.message.text
    context.user_data["goals"] = goals

    # Переходим к следующему шагу
    await update.message.reply_text("Отлично! Теперь укажите, какие ресурсы у вас уже есть (например, книги, курсы, наставники).")
    return WAITING_FOR_RESOURCES  # Переходим к состоянию ожидания ресурсов

# Обработчик для получения ресурсов
async def handle_resources(update: Update, context: CallbackContext):
    resources = update.message.text
    context.user_data["resources"] = resources

    # Генерация плана с помощью модели
    goals = context.user_data.get("goals", "не указаны")
    resources = context.user_data.get("resources", "не указаны")

    # Формируем промт для модели
    prompt = (
        f"Пользователь хочет улучшить следующие области: {goals}. "
        f"У него есть следующие ресурсы: {resources}. "
        f"Составь подробный план саморазвития, включающий конкретные шаги и рекомендации."
    )

    # Генерация плана с помощью модели
    plan = generate_response(prompt)

    # Отправка плана пользователю
    await update.message.reply_text(plan, reply_markup=get_keyboard())

    # Завершаем диалог
    return ConversationHandler.END

# Обработчик отмены диалога
async def cancel(update: Update, context: CallbackContext):
    await update.message.reply_text("Диалог отменен.", reply_markup=get_keyboard())
    return ConversationHandler.END

import sqlite3
from telegram.ext import ConversationHandler

# Определяем состояния диалога для регистрации
WAITING_FOR_NAME, WAITING_FOR_AGE, WAITING_FOR_INTERESTS, WAITING_FOR_PROBLEMS = range(4)

# Инициализация базы данных
def init_db():
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS users (
            user_id INTEGER PRIMARY KEY,
            username TEXT,
            first_name TEXT,
            last_name TEXT,
            age INTEGER,
            interests TEXT,
            problems TEXT,
            registered_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
        )
    ''')
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS goals (
            goal_id INTEGER PRIMARY KEY AUTOINCREMENT,
            user_id INTEGER,
            description TEXT,
            deadline TEXT,
            progress INTEGER DEFAULT 0,
            status TEXT DEFAULT 'В процессе',
            FOREIGN KEY (user_id) REFERENCES users (user_id)
        )
    ''')
    conn.commit()
    conn.close()

# Регистрация пользователя
def register_user(user_id, username, first_name, last_name, age, interests, problems):
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    cursor.execute('''
        INSERT OR REPLACE INTO users (user_id, username, first_name, last_name, age, interests, problems)
        VALUES (?, ?, ?, ?, ?, ?, ?)
    ''', (user_id, username, first_name, last_name, age, interests, problems))
    conn.commit()
    conn.close()

# Получение информации о пользователе
def get_user_info(user_id):
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    cursor.execute('SELECT * FROM users WHERE user_id = ?', (user_id,))
    user_info = cursor.fetchone()
    conn.close()
    return user_info

# Обработчик команды /start
async def start(update: Update, context: CallbackContext):
    user = update.effective_user
    user_info = get_user_info(user.id)

    # Если пользователь уже зарегистрирован
    if user_info:
        await update.message.reply_text(
            f"👋 Привет, {user_info[2]}! Рад снова тебя видеть. Чем могу помочь?",
            reply_markup=get_keyboard()
        )
        return ConversationHandler.END

    # Если пользователь не зарегистрирован
    await update.message.reply_text("👋 Привет! Давай познакомимся. Как тебя зовут?")
    return WAITING_FOR_NAME

# Обработчик для получения имени
async def handle_name(update: Update, context: CallbackContext):
    user = update.effective_user
    context.user_data["first_name"] = update.message.text
    await update.message.reply_text("Приятно познакомиться! Сколько тебе лет?")
    return WAITING_FOR_AGE

# Обработчик для получения возраста
async def handle_age(update: Update, context: CallbackContext):
    try:
        age = int(update.message.text)
        if age < 0 or age > 120:
            await update.message.reply_text("Пожалуйста, введите корректный возраст (от 0 до 120).")
            return WAITING_FOR_AGE
        context.user_data["age"] = age
        await update.message.reply_text("Отлично! Какие у тебя интересы? (например, спорт, программирование, музыка)")
        return WAITING_FOR_INTERESTS
    except ValueError:
        await update.message.reply_text("Пожалуйста, введите возраст числом.")
        return WAITING_FOR_AGE

# Обработчик для получения интересов
async def handle_interests(update: Update, context: CallbackContext):
    context.user_data["interests"] = update.message.text
    await update.message.reply_text("Хорошо! Какие у тебя основные проблемы или вопросы, с которыми ты хотел бы помочь?")
    return WAITING_FOR_PROBLEMS

# Обработчик для получения проблем
async def handle_problems(update: Update, context: CallbackContext):
    user = update.effective_user
    problems = update.message.text
    context.user_data["problems"] = problems

    # Сохраняем данные пользователя в базу данных
    register_user(
        user.id,
        user.username,
        context.user_data.get("first_name"),
        user.last_name,
        context.user_data.get("age"),
        context.user_data.get("interests"),
        problems
    )

    await update.message.reply_text(
        "Спасибо за регистрацию! Теперь я знаю о тебе немного больше. Чем могу помочь?",
        reply_markup=get_keyboard()
    )
    return ConversationHandler.END

# Обработчик отмены диалога
async def cancel(update: Update, context: CallbackContext):
    await update.message.reply_text("Регистрация отменена.", reply_markup=get_keyboard())
    return ConversationHandler.END

# Клавиатура с кнопками
def get_keyboard():
    keyboard = [
        ["Задать вопрос🤔", "Профиль"],
        ["Наш канал!", "Поддержка😌", "Тест"]
    ]
    return ReplyKeyboardMarkup(keyboard, resize_keyboard=True)

# Обработчик для кнопки "Профиль"
async def show_profile(update: Update, context: CallbackContext):
    user = update.effective_user
    user_info = get_user_info(user.id)

    # Если пользователь зарегистрирован
    if user_info:
        profile_message = (
            f"👤 Ваш профиль:\n\n"
            f"Имя: {user_info[2]}\n"
            f"Возраст: {user_info[4]}\n"
            f"Интересы: {user_info[5]}\n"
            f"Основные проблемы: {user_info[6]}"
        )
        await update.message.reply_text(profile_message, reply_markup=get_keyboard())
    else:
        await update.message.reply_text("Вы еще не зарегистрированы. Нажмите /start, чтобы начать.", reply_markup=get_keyboard())

async def handle_message(update: Update, context: CallbackContext):
    user_message = update.message.text
    user_data = context.user_data

    # Если пользователь находится в состоянии добавления цели
    if user_data.get("state") in ["WAITING_FOR_GOAL_DESCRIPTION", "WAITING_FOR_GOAL_DEADLINE"]:
        # Передаем управление обработчикам целей
        return await handle_goal_description(update, context) if user_data.get("state") == "WAITING_FOR_GOAL_DESCRIPTION" else await handle_goal_deadline(update, context)

    # Если бот ждёт имя пользователя
    if user_data.get("state") == "waiting_for_name":
        user_data["user_name"] = user_message  # Сохраняем имя пользователя
        user_data["state"] = None  # Сбрасываем состояние
        response = f"Приятно познакомиться, {user_message}! Чем могу помочь?"
        await update.message.reply_text(response)
        await text_to_speech(update, response)  # Отправка голосового ответа
        return

    # Обработка кнопок
    if user_message == "Тест":
        await start_test(update, context)
    elif user_message == "Задать вопрос🤔":
        response = "Пожалуйста, задайте ваш вопрос текстом или отправьте голосовое сообщение."
        await update.message.reply_text(response)
        await text_to_speech(update, response)  # Отправка голосового ответа
    elif user_message == "Наш канал!":
        response = "Подписывайтесь на наш канал: https://t.me/mentachannel"
        await update.message.reply_text(response)
        await text_to_speech(update, response)  # Отправка голосового ответа
    elif user_message == "Поддержка😌":
        response = "Свяжитесь с поддержкой: @adinaadinaadina"
        await update.message.reply_text(response)
        await text_to_speech(update, response)  # Отправка голосового ответа
    elif user_message == "Профиль":
        await show_profile(update, context)
    elif user_message == "Добавить цель":  # Обработка кнопки "Добавить цель"
        await add_goal(update, context)
    elif user_message == "Мои цели":  # Обработка кнопки "Мои цели"
        await show_goals(update, context)
    elif user_message == "Отметить прогресс":  # Обработка кнопки "Отметить прогресс"
        await update_progress(update, context)
    elif user_message == "Удалить цель":  # Обработка кнопки "Удалить цель"
        await delete_goal(update, context)
    else:
        # Проверка на "понятность" запроса
        if not is_message_clear(user_message):
            response = (
                "Извините, я не совсем понял ваш запрос. Пожалуйста, уточните, что вы имеете в виду?\n\n"
                "Например, вы можете спросить:\n"
                "- Как подготовиться к экзаменам?\n"
                "- Что делать при финансовой нестабильности?\n"
                "- Как найти свое призвание?"
            )
            await update.message.reply_text(response)
            await text_to_speech(update, response)  # Отправка голосового ответа
            return

        # Показываем индикатор печати
        await send_typing_action(update)

        # Определяем callback для отправки ответа
        async def send_text_response(response):
            await update.message.reply_text(response)
            await text_to_speech(update, response)  # Отправка голосового ответа

        # Добавляем запрос в очередь
        user_id = update.effective_user.id
        user_name = user_data.get("user_name")
        message_timestamps[user_id] = time.time()
        message_queue.put({
            "user_id": user_id,
            "question": user_message,
            "callback": send_text_response,
            "user_name": user_name
        })

from telegram import ReplyKeyboardMarkup

# Клавиатура для управления целями
def get_goals_keyboard():
    keyboard = [
        ["Добавить цель", "Мои цели"],
        ["Отметить прогресс", "Удалить цель"],
        ["Назад"]
    ]
    return ReplyKeyboardMarkup(keyboard, resize_keyboard=True)

# Обработчик команды /goals
async def goals(update: Update, context: CallbackContext):
    await update.message.reply_text(
        "🎯 Управление целями. Выберите действие:",
        reply_markup=get_goals_keyboard()
    )


async def add_goal(update: Update, context: CallbackContext):
    await update.message.reply_text("Напишите описание вашей цели:")
    context.user_data["state"] = "WAITING_FOR_GOAL_DESCRIPTION"  # Устанавливаем состояние
    return "WAITING_FOR_GOAL_DESCRIPTION"

async def handle_goal_description(update: Update, context: CallbackContext):
    description = update.message.text
    context.user_data["goal_description"] = description
    await update.message.reply_text("Укажите срок выполнения цели (например, 2023-12-31):")
    context.user_data["state"] = "WAITING_FOR_GOAL_DEADLINE"  # Устанавливаем состояние
    return "WAITING_FOR_GOAL_DEADLINE"

async def handle_goal_deadline(update: Update, context: CallbackContext):
    deadline = update.message.text
    user_id = update.effective_user.id
    description = context.user_data["goal_description"]

    # Сохраняем цель в базу данных
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    cursor.execute('''
        INSERT INTO goals (user_id, description, deadline)
        VALUES (?, ?, ?)
    ''', (user_id, description, deadline))
    conn.commit()
    conn.close()

    await update.message.reply_text("Цель успешно добавлена!", reply_markup=get_goals_keyboard())
    context.user_data["state"] = None  # Сбрасываем состояние
    return ConversationHandler.END


async def show_goals(update: Update, context: CallbackContext):
    user_id = update.effective_user.id
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    cursor.execute('SELECT description, deadline, progress, status FROM goals WHERE user_id = ?', (user_id,))
    goals = cursor.fetchall()
    conn.close()

    if not goals:
        await update.message.reply_text("У вас пока нет целей.", reply_markup=get_goals_keyboard())
        return

    goals_message = "🎯 Ваши цели:\n\n"
    for goal in goals:
        goals_message += f"• {goal[0]}\nСрок: {goal[1]}\nПрогресс: {goal[2]}%\nСтатус: {goal[3]}\n\n"

    await update.message.reply_text(goals_message, reply_markup=get_goals_keyboard())


async def update_progress(update: Update, context: CallbackContext):
    user_id = update.effective_user.id
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    cursor.execute('SELECT goal_id, description FROM goals WHERE user_id = ?', (user_id,))
    goals = cursor.fetchall()
    conn.close()

    if not goals:
        await update.message.reply_text("У вас пока нет целей.", reply_markup=get_goals_keyboard())
        return

    keyboard = [
        [InlineKeyboardButton(goal[1], callback_data=f"progress_{goal[0]}")]
        for goal in goals
    ]
    reply_markup = InlineKeyboardMarkup(keyboard)
    await update.message.reply_text("Выберите цель для обновления прогресса:", reply_markup=reply_markup)

async def check_goal_completion(update: Update, context: CallbackContext):
    user_id = update.effective_user.id
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    cursor.execute('SELECT goal_id, description FROM goals WHERE user_id = ?', (user_id,))
    goals = cursor.fetchall()
    conn.close()

    if not goals:
        await update.message.reply_text("У вас пока нет целей.", reply_markup=get_goals_keyboard())
        return

    # Создаем инлайн-кнопки для каждой цели
    keyboard = [
        [InlineKeyboardButton(goal[1], callback_data=f"complete_{goal[0]}")]
        for goal in goals
    ]
    reply_markup = InlineKeyboardMarkup(keyboard)
    await update.message.reply_text("Выберите цель для проверки выполнения:", reply_markup=reply_markup)


async def handle_goal_selection(update: Update, context: CallbackContext):
    query = update.callback_query
    await query.answer()

    # Извлекаем goal_id из callback_data
    goal_id = int(query.data.replace("complete_", ""))

    # Сохраняем goal_id в user_data для использования в следующем шаге
    context.user_data["goal_id"] = goal_id

    # Запрашиваем, выполнена ли цель
    await query.edit_message_text("Цель выполнена? (Да/Нет):")
    return "WAITING_FOR_COMPLETION_STATUS"


async def handle_completion_status(update: Update, context: CallbackContext):
    user_response = update.message.text.lower()

    # Проверяем ответ пользователя
    if user_response not in ["да", "нет"]:
        await update.message.reply_text("Пожалуйста, ответьте 'Да' или 'Нет'.")
        return "WAITING_FOR_COMPLETION_STATUS"

    # Извлекаем goal_id из user_data
    goal_id = context.user_data.get("goal_id")
    if not goal_id:
        await update.message.reply_text("Ошибка: цель не найдена. Попробуйте снова.")
        return ConversationHandler.END

    # Обновляем статус цели в базе данных
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    if user_response == "да":
        cursor.execute('UPDATE goals SET status = ? WHERE goal_id = ?', ("Завершено", goal_id))
        await update.message.reply_text("Цель отмечена как выполненная!", reply_markup=get_goals_keyboard())
    else:
        cursor.execute('UPDATE goals SET status = ? WHERE goal_id = ?', ("В процессе", goal_id))
        await update.message.reply_text("Цель остается в процессе выполнения.", reply_markup=get_goals_keyboard())
    conn.commit()
    conn.close()

    return ConversationHandler.END


import schedule
import time
from threading import Thread

# Функция для отправки отчетов
def send_reports():
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    cursor.execute('SELECT user_id FROM users')
    users = cursor.fetchall()
    conn.close()

    for user in users:
        user_id = user[0]
        conn = sqlite3.connect('users.db')
        cursor = conn.cursor()
        cursor.execute('SELECT description, progress FROM goals WHERE user_id = ?', (user_id,))
        goals = cursor.fetchall()
        conn.close()

        if goals:
            report_message = "📊 Ваш еженедельный отчет по целям:\n\n"
            for goal in goals:
                report_message += f"• {goal[0]}\nПрогресс: {goal[1]}%\n\n"

            # Отправляем отчет пользователю
            context.bot.send_message(chat_id=user_id, text=report_message)

# Запуск планировщика для отчетов
def start_scheduler():
    schedule.every().monday.at("09:00").do(send_reports)  # Еженедельно в понедельник в 9:00
    while True:
        schedule.run_pending()
        time.sleep(1)

# Запуск планировщика в отдельном потоке
scheduler_thread = Thread(target=start_scheduler, daemon=True)
scheduler_thread.start()




def main():
    # Инициализация базы данных
    init_db()

    # Запуск обработчика очереди сообщений в отдельном потоке
    queue_thread = threading.Thread(target=message_queue_worker, daemon=True)
    queue_thread.start()

    application = Application.builder().token(TOKEN).build()

    # Set bot commands in menu
    commands = [
        ("start", "Перезапустить бота")
    ]

    # Set commands when bot starts
    async def post_init(application: Application):
        await application.bot.set_my_commands(commands)

    application.post_init = post_init

    # Регистрация ConversationHandler для регистрации пользователя
    conv_handler = ConversationHandler(
        entry_points=[CommandHandler("start", start)],
        states={
            WAITING_FOR_NAME: [MessageHandler(filters.TEXT & ~filters.COMMAND, handle_name)],
            WAITING_FOR_AGE: [MessageHandler(filters.TEXT & ~filters.COMMAND, handle_age)],
            WAITING_FOR_INTERESTS: [MessageHandler(filters.TEXT & ~filters.COMMAND, handle_interests)],
            WAITING_FOR_PROBLEMS: [MessageHandler(filters.TEXT & ~filters.COMMAND, handle_problems)],
        },
        fallbacks=[CommandHandler("cancel", cancel)],
    )

    # Регистрация ConversationHandler для управления целями
    # Регистрация ConversationHandler для управления целями
    goals_handler = ConversationHandler(
        entry_points=[CommandHandler("goals", goals)],
        states={
            "WAITING_FOR_GOAL_DESCRIPTION": [MessageHandler(filters.TEXT & ~filters.COMMAND, handle_goal_description)],
            "WAITING_FOR_GOAL_DEADLINE": [MessageHandler(filters.TEXT & ~filters.COMMAND, handle_goal_deadline)],
            "WAITING_FOR_COMPLETION_STATUS": [MessageHandler(filters.TEXT & ~filters.COMMAND, handle_completion_status)],
        },
        fallbacks=[CommandHandler("cancel", cancel)],
    )
    application.add_handler(goals_handler)

# Регистрация обработчика для выбора цели
    application.add_handler(CallbackQueryHandler(handle_goal_selection, pattern="^complete_"))


    # Регистрация обработчиков
    application.add_handler(conv_handler)
    application.add_handler(CommandHandler("goals", goals))
    application.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, handle_message))
    application.add_handler(MessageHandler(filters.VOICE, handle_voice))
    application.add_handler(CallbackQueryHandler(handle_test_selection, pattern="^test_"))
    application.add_handler(CallbackQueryHandler(handle_test_answer, pattern="^answer_"))

    # Запуск бота
    application.run_polling()

if __name__ == '__main__':
    main()
```
