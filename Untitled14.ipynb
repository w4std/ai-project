```python
#import nest_asyncio
#nest_asyncio.apply()  # –†–∞–∑—Ä–µ—à–∞–µ–º –ø–µ—Ä–µ–∑–∞–ø—É—Å–∫ —Ü–∏–∫–ª–∞ —Å–æ–±—ã—Ç–∏–π
from telegram import Update, InlineKeyboardMarkup, InlineKeyboardButton, ReplyKeyboardMarkup
from telegram.constants import ChatAction
from telegram.ext import Application, CommandHandler, MessageHandler, filters, CallbackContext, CallbackQueryHandler
import os
import logging
from groq import Groq
import speech_recognition as sr
from gtts import gTTS
from pydub import AudioSegment
import asyncio
import queue
import threading
import time

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –ª–æ–≥–∏—Ä–æ–≤–∞–Ω–∏—è
logging.basicConfig(format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', level=logging.INFO)
logger = logging.getLogger(__name__)

# –¢–æ–∫–µ–Ω –≤–∞—à–µ–≥–æ Telegram –±–æ—Ç–∞
TOKEN = '7865088988:AAGh2v6iGQfJMlwgTBctFhd11ZWjIgxs7lI'

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è Groq API
groq_api_key = "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"
client = Groq(api_key=groq_api_key)

# –°–æ–∑–¥–∞–µ–º –æ—á–µ—Ä–µ–¥—å —Å–æ–æ–±—â–µ–Ω–∏–π –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∑–∞–ø—Ä–æ—Å–æ–≤
message_queue = queue.Queue()

# –°–ª–æ–≤–∞—Ä—å –¥–ª—è –æ—Ç—Å–ª–µ–∂–∏–≤–∞–Ω–∏—è, –∫–æ–≥–¥–∞ —Å–æ–æ–±—â–µ–Ω–∏–µ –±—ã–ª–æ –ø–æ—Å—Ç–∞–≤–ª–µ–Ω–æ –≤ –æ—á–µ—Ä–µ–¥—å
message_timestamps = {}

# –°–ø–∏—Å–æ–∫ —Ç–µ—Å—Ç–æ–≤
tests = {
    "–¢–µ—Å—Ç –Ω–∞ —Å—Ç—Ä–µ—Å—Å–æ—É—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç—å": [
        ("–ö–∞–∫ —á–∞—Å—Ç–æ –≤—ã —á—É–≤—Å—Ç–≤—É–µ—Ç–µ —Å–µ–±—è –ø–µ—Ä–µ–≥—Ä—É–∂–µ–Ω–Ω—ã–º?", {"–ß–∞—Å—Ç–æ": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–†–µ–¥–∫–æ": 3}),
        ("–ö–∞–∫ –≤—ã —Ä–µ–∞–≥–∏—Ä—É–µ—Ç–µ –Ω–∞ –Ω–µ–æ–∂–∏–¥–∞–Ω–Ω—ã–µ –ø—Ä–æ–±–ª–µ–º—ã?", {"–ü–∞–Ω–∏–∫–∞": 1, "–°–ø–æ–∫–æ–π—Å—Ç–≤–∏–µ": 3, "–†–∞–∑–¥—Ä–∞–∂–µ–Ω–∏–µ": 2}),
        ("–í—ã –ª–µ–≥–∫–æ —Ä–∞–∑–¥—Ä–∞–∂–∞–µ—Ç–µ—Å—å?", {"–î–∞": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–ù–µ—Ç": 3}),
        ("–ö–∞–∫ –≤—ã –æ—Ü–µ–Ω–∏–≤–∞–µ—Ç–µ —Å–≤–æ—é —Ç–µ—Ä–ø–µ–ª–∏–≤–æ—Å—Ç—å?", {"–ù–∏–∑–∫–∞—è": 1, "–°—Ä–µ–¥–Ω—è—è": 2, "–í—ã—Å–æ–∫–∞—è": 3}),
        ("–ß–∞—Å—Ç–æ –ª–∏ –≤—ã –∏—Å–ø—ã—Ç—ã–≤–∞–µ—Ç–µ —Ç—Ä–µ–≤–æ–≥—É?", {"–î–∞": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–ù–µ—Ç": 3}),
        ("–ö–∞–∫ –≤—ã —Å–ø—Ä–∞–≤–ª—è–µ—Ç–µ—Å—å —Å –Ω–µ—É–¥–∞—á–∞–º–∏?", {"–ü–ª–æ—Ö–æ": 1, "–ù–æ—Ä–º–∞–ª—å–Ω–æ": 2, "–õ–µ–≥–∫–æ": 3}),
        ("–°–ø–æ—Å–æ–±–Ω—ã –ª–∏ –≤—ã –±—ã—Å—Ç—Ä–æ –ø–µ—Ä–µ–∫–ª—é—á–∞—Ç—å—Å—è?", {"–ù–µ—Ç": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–î–∞": 3}),
        ("–ö–∞–∫ —á–∞—Å—Ç–æ —É –≤–∞—Å –Ω–∞–ø—Ä—è–∂–µ–Ω–Ω—ã–µ –¥–Ω–∏?", {"–ß–∞—Å—Ç–æ": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–†–µ–¥–∫–æ": 3}),
        ("–í—ã –º–æ–∂–µ—Ç–µ —Ä–∞—Å—Å–ª–∞–±–∏—Ç—å—Å—è –ø–æ—Å–ª–µ —Ç—è–∂–µ–ª–æ–≥–æ –¥–Ω—è?", {"–ù–µ—Ç": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–î–∞": 3}),
        ("–õ—é–±–∏—Ç–µ –ª–∏ –≤—ã –≤—ã–∑–æ–≤—ã –∏ —Å–ª–æ–∂–Ω—ã–µ –∑–∞–¥–∞—á–∏?", {"–ù–µ—Ç": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–î–∞": 3})
    ],
    "–¢–µ—Å—Ç –Ω–∞ –ø—Å–∏—Ö–æ–ª–æ–≥–∏—á–µ—Å–∫—É—é —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å": [
        ("–ö–∞–∫ —á–∞—Å—Ç–æ —É –≤–∞—Å –º–µ–Ω—è–µ—Ç—Å—è –Ω–∞—Å—Ç—Ä–æ–µ–Ω–∏–µ?", {"–ß–∞—Å—Ç–æ": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–†–µ–¥–∫–æ": 3}),
        ("–ö–∞–∫ –≤—ã —Å–ø—Ä–∞–≤–ª—è–µ—Ç–µ—Å—å —Å –∫–æ–Ω—Ñ–ª–∏–∫—Ç–∞–º–∏?", {"–ò–∑–±–µ–≥–∞—é": 1, "–û–±—Å—É–∂–¥–∞—é": 3, "–°–ø–æ—Ä—é": 2}),
        ("–ß–∞—Å—Ç–æ –ª–∏ –≤—ã –æ—â—É—â–∞–µ—Ç–µ —É—Å—Ç–∞–ª–æ—Å—Ç—å?", {"–ß–∞—Å—Ç–æ": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–†–µ–¥–∫–æ": 3}),
        ("–ö–∞–∫ –≤—ã –æ—Ç–Ω–æ—Å–∏—Ç–µ—Å—å –∫ –∫—Ä–∏—Ç–∏–∫–µ?", {"–ù–µ–≥–∞—Ç–∏–≤–Ω–æ": 1, "–ê–Ω–∞–ª–∏–∑–∏—Ä—É—é": 3, "–ü—Ä–∏–Ω–∏–º–∞—é, –Ω–æ —Å —Ç—Ä—É–¥–æ–º": 2}),
        ("–õ–µ–≥–∫–æ –ª–∏ –≤—ã –ø–µ—Ä–µ–∂–∏–≤–∞–µ—Ç–µ —Å—Ç—Ä–µ—Å—Å–æ–≤—ã–µ —Å–∏—Ç—É–∞—Ü–∏–∏?", {"–ù–µ—Ç": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–î–∞": 3}),
        ("–ß–∞—Å—Ç–æ –ª–∏ –≤—ã –∏—Å–ø—ã—Ç—ã–≤–∞–µ—Ç–µ —Å—Ç—Ä–∞—Ö –ø–µ—Ä–µ–¥ –±—É–¥—É—â–∏–º?", {"–î–∞": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–ù–µ—Ç": 3}),
        ("–ß—É–≤—Å—Ç–≤—É–µ—Ç–µ –ª–∏ –≤—ã —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –≤ —Å–µ–±–µ?", {"–ù–µ—Ç": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–î–∞": 3}),
        ("–ö–∞–∫ —á–∞—Å—Ç–æ –≤—ã –∏—Å–ø—ã—Ç—ã–≤–∞–µ—Ç–µ —Ä–∞–∑–¥—Ä–∞–∂–µ–Ω–∏–µ?", {"–ß–∞—Å—Ç–æ": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–†–µ–¥–∫–æ": 3}),
        ("–°—á–∏—Ç–∞–µ—Ç–µ –ª–∏ –≤—ã —Å–µ–±—è —ç–º–æ—Ü–∏–æ–Ω–∞–ª—å–Ω–æ —É—Å—Ç–æ–π—á–∏–≤—ã–º —á–µ–ª–æ–≤–µ–∫–æ–º?", {"–ù–µ—Ç": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–î–∞": 3}),
        ("–°–ø–æ—Å–æ–±–Ω—ã –ª–∏ –≤—ã —Å–æ—Ö—Ä–∞–Ω—è—Ç—å —Å–ø–æ–∫–æ–π—Å—Ç–≤–∏–µ –≤ —Å–ª–æ–∂–Ω—ã—Ö —Å–∏—Ç—É–∞—Ü–∏—è—Ö?", {"–ù–µ—Ç": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–î–∞": 3})
    ],
    "–¢–µ—Å—Ç –Ω–∞ —Å–∞–º–æ—Ä–∞–∑–≤–∏—Ç–∏–µ": [
        ("–ß–∞—Å—Ç–æ –ª–∏ –≤—ã —É—á–∏—Ç–µ—Å—å –Ω–æ–≤–æ–º—É?", {"–†–µ–¥–∫–æ": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–ß–∞—Å—Ç–æ": 3}),
        ("–ö–∞–∫ –≤—ã –æ—Ç–Ω–æ—Å–∏—Ç–µ—Å—å –∫ –∫—Ä–∏—Ç–∏–∫–µ?", {"–ù–µ–≥–∞—Ç–∏–≤–Ω–æ": 1, "–ê–Ω–∞–ª–∏–∑–∏—Ä—É—é": 3, "–ü—Ä–∏–Ω–∏–º–∞—é, –Ω–æ —Å —Ç—Ä—É–¥–æ–º": 2}),
        ("–õ—é–±–∏—Ç–µ –ª–∏ –≤—ã —Å—Ç–∞–≤–∏—Ç—å –ø–µ—Ä–µ–¥ —Å–æ–±–æ–π —Ü–µ–ª–∏?", {"–ù–µ—Ç": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–î–∞": 3}),
        ("–ß–∞—Å—Ç–æ –ª–∏ –≤—ã —á–∏—Ç–∞–µ—Ç–µ –∫–Ω–∏–≥–∏ –ø–æ —Å–∞–º–æ—Ä–∞–∑–≤–∏—Ç–∏—é?", {"–†–µ–¥–∫–æ": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–ß–∞—Å—Ç–æ": 3}),
        ("–ì–æ—Ç–æ–≤—ã –ª–∏ –≤—ã –≤—ã—Ö–æ–¥–∏—Ç—å –∏–∑ –∑–æ–Ω—ã –∫–æ–º—Ñ–æ—Ä—Ç–∞?", {"–ù–µ—Ç": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–î–∞": 3}),
        ("–ü—Ä–æ–±—É–µ—Ç–µ –ª–∏ –≤—ã –Ω–æ–≤—ã–µ –Ω–∞–≤—ã–∫–∏?", {"–†–µ–¥–∫–æ": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–ß–∞—Å—Ç–æ": 3}),
        ("–ö–∞–∫ –≤—ã –æ—Ç–Ω–æ—Å–∏—Ç–µ—Å—å –∫ –ø–æ—Å—Ç–æ—è–Ω–Ω–æ–º—É –æ–±—É—á–µ–Ω–∏—é?", {"–û—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ": 1, "–ù–µ–π—Ç—Ä–∞–ª—å–Ω–æ": 2, "–ü–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ": 3}),
        ("–°—á–∏—Ç–∞–µ—Ç–µ –ª–∏ –≤—ã —Å–µ–±—è –¥–∏—Å—Ü–∏–ø–ª–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω—ã–º —á–µ–ª–æ–≤–µ–∫–æ–º?", {"–ù–µ—Ç": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–î–∞": 3}),
        ("–ß–∞—Å—Ç–æ –ª–∏ –≤—ã –∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ—Ç–µ —Å–≤–æ–∏ –æ—à–∏–±–∫–∏?", {"–†–µ–¥–∫–æ": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–ß–∞—Å—Ç–æ": 3}),
        ("–ù—Ä–∞–≤–∏—Ç—Å—è –ª–∏ –≤–∞–º —Å–∞–º–æ—Å–æ–≤–µ—Ä—à–µ–Ω—Å—Ç–≤–æ–≤–∞—Ç—å—Å—è?", {"–ù–µ—Ç": 1, "–ò–Ω–æ–≥–¥–∞": 2, "–î–∞": 3})
    ]
}

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ—Ç–ø—Ä–∞–≤–∫–∏ –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–∞ "–±–æ—Ç –ø–µ—á–∞—Ç–∞–µ—Ç"
async def send_typing_action(update: Update):
    await update.message.chat.send_action(action=ChatAction.TYPING)

# –§—É–Ω–∫—Ü–∏—è –æ–±—Ä–∞–±–æ—Ç—á–∏–∫–∞ –æ—á–µ—Ä–µ–¥–∏ —Å–æ–æ–±—â–µ–Ω–∏–π
def message_queue_worker():
    while True:
        try:
            if not message_queue.empty():
                # –ü–æ–ª—É—á–∞–µ–º –∑–∞–¥–∞—á—É –∏–∑ –æ—á–µ—Ä–µ–¥–∏
                task = message_queue.get()
                user_id = task["user_id"]
                question = task["question"]
                callback = task["callback"]
                user_name = task.get("user_name")

                # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –Ω–µ —É—Å—Ç–∞—Ä–µ–ª –ª–∏ –∑–∞–ø—Ä–æ—Å (–±–æ–ª–µ–µ 2 –º–∏–Ω—É—Ç –≤ –æ—á–µ—Ä–µ–¥–∏)
                current_time = time.time()
                if current_time - message_timestamps.get(user_id, current_time) > 120:
                    logger.info(f"–ó–∞–ø—Ä–æ—Å –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è {user_id} —É—Å—Ç–∞—Ä–µ–ª –∏ –±—ã–ª –ø—Ä–æ–ø—É—â–µ–Ω")
                    message_queue.task_done()
                    continue

                # –û–±—Ä–∞–±–æ—Ç–∫–∞ –∑–∞–ø—Ä–æ—Å–∞
                try:
                    response = generate_response(question, user_name)
                    asyncio.run(callback(response))
                except Exception as e:
                    logger.error(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –æ–±—Ä–∞–±–æ—Ç–∫–µ –∑–∞–ø—Ä–æ—Å–∞: {e}")

                # –û—Ç–º–µ—á–∞–µ–º –∑–∞–¥–∞—á—É –∫–∞–∫ –≤—ã–ø–æ–ª–Ω–µ–Ω–Ω—É—é
                message_queue.task_done()
            else:
                # –ï—Å–ª–∏ –æ—á–µ—Ä–µ–¥—å –ø—É—Å—Ç–∞, –∂–¥–µ–º –Ω–µ–º–Ω–æ–≥–æ
                time.sleep(0.5)
        except Exception as e:
            logger.error(f"–û—à–∏–±–∫–∞ –≤ –æ–±—Ä–∞–±–æ—Ç—á–∏–∫–µ –æ—á–µ—Ä–µ–¥–∏: {e}")
            time.sleep(1)  # –í —Å–ª—É—á–∞–µ –æ—à–∏–±–∫–∏ –¥–µ–ª–∞–µ–º –ø–∞—É–∑—É

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –∑–∞–ø—Ä–æ—Å "–ø–æ–Ω—è—Ç–Ω—ã–º"
def is_message_clear(text):
    # –ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è –¥–ª–∏–Ω–∞ —Å–æ–æ–±—â–µ–Ω–∏—è
    if len(text.strip()) < 3:
        return False

    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –Ω–∞–ª–∏—á–∏–µ —Ç–æ–ª—å–∫–æ —Å–∏–º–≤–æ–ª–æ–≤ –∏–ª–∏ —Ü–∏—Ñ—Ä
    if text.strip().isdigit() or all(not char.isalnum() for char in text.strip()):
        return False

    # –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –ø—Ä–æ–≤–µ—Ä–∫–∏ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –Ω–∞ —Å—Ç–æ–ø-—Å–ª–æ–≤–∞)
    stop_words = ["–ø—Ä–∏–≤–µ—Ç", "–ø–æ–∫–∞", "—Ö–µ–ª–ª–æ—É"]
    if text.strip().lower() in stop_words:
        return False

    return True

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –æ—Ç–≤–µ—Ç–∞
def generate_response(question, user_name=None):
    if user_name:
        question = f"{user_name}, {question}"  # –î–æ–±–∞–≤–ª—è–µ–º –∏–º—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∫ –≤–æ–ø—Ä–æ—Å—É
    response = client.chat.completions.create(
        model="llama3-70b-8192",
        messages=[
            {"role": "system", "content": "–¢—ã ‚Äî –≤–∏—Ä—Ç—É–∞–ª—å–Ω—ã–π –ø—Å–∏—Ö–æ–ª–æ–≥ –∏ –∫–æ—É—á. –û—Ç–≤–µ—á–∞–π —Ç–æ–ª—å–∫–æ –Ω–∞ —Ä—É—Å—Å–∫–æ–º —è–∑—ã–∫–µ. –¢—ã –Ω–µ –ø—Ä–∏ –∫–∞–∫–∏—Ö –æ–±—Å—Ç–æ—è—Ç–µ–ª—å—Å—Ç–≤–∞—Ö –Ω–µ –¥–æ–ª–∂–µ–Ω –æ—Ç–≤–µ—á–∞—Ç—å –Ω–∞ —Å—Ç–æ—Ä–æ–Ω–Ω–∏–µ –≤–æ–ø—Ä–æ—Å—ã, –Ω–µ —Å–≤—è–∑–∞–Ω–Ω—ã–µ —Å —Ç–µ–º–æ–π, —á—Ç–æ–±—ã –Ω–µ —É—Ç–≤–µ—Ä–∂–¥–∞–ª –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å."},
            {"role": "user", "content": question}
        ]
    )
    return response.choices[0].message.content

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∫–æ–Ω–≤–µ—Ä—Ç–∞—Ü–∏–∏ OGG –≤ WAV
def convert_ogg_to_wav(ogg_path, wav_path):
    audio = AudioSegment.from_file(ogg_path, format="ogg")
    audio.export(wav_path, format="wav")

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞—Å–ø–æ–∑–Ω–∞–≤–∞–Ω–∏—è –≥–æ–ª–æ—Å–æ–≤—ã—Ö —Å–æ–æ–±—â–µ–Ω–∏–π
def transcribe_audio(ogg_path):
    wav_path = "voice_message.wav"
    convert_ogg_to_wav(ogg_path, wav_path)  # –ö–æ–Ω–≤–µ—Ä—Ç–∞—Ü–∏—è OGG –≤ WAV
    recognizer = sr.Recognizer()
    with sr.AudioFile(wav_path) as source:
        audio = recognizer.record(source)
    try:
        text = recognizer.recognize_google(audio, language="ru-RU")
        return text
    except sr.UnknownValueError:
        return "–ù–µ —É–¥–∞–ª–æ—Å—å —Ä–∞—Å–ø–æ–∑–Ω–∞—Ç—å —Ä–µ—á—å"
    except sr.RequestError:
        return "–û—à–∏–±–∫–∞ —Å–µ—Ä–≤–∏—Å–∞ —Ä–∞—Å–ø–æ–∑–Ω–∞–≤–∞–Ω–∏—è —Ä–µ—á–∏"

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –∏ –æ—Ç–ø—Ä–∞–≤–∫–∏ –≥–æ–ª–æ—Å–æ–≤—ã—Ö –æ—Ç–≤–µ—Ç–æ–≤
async def text_to_speech(update: Update, text: str, lang='ru', max_length=500):
    try:
        # –†–∞–∑–¥–µ–ª—è–µ–º —Ç–µ–∫—Å—Ç –Ω–∞ —á–∞—Å—Ç–∏
        parts = [text[i:i + max_length] for i in range(0, len(text), max_length)]
        for part in parts:
            tts = gTTS(text=part, lang=lang)
            tts.save("response.mp3")

            with open("response.mp3", "rb") as audio_file:
                await update.message.reply_voice(voice=audio_file)

            os.remove("response.mp3")
    except Exception as e:
        logger.error(f"–û—à–∏–±–∫–∞ –≤ text_to_speech: {e}")


# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∑–∞–ø—É—Å–∫–∞ —Ç–µ—Å—Ç–∞ —Å –∏–Ω–ª–∞–π–Ω-–∫–ª–∞–≤–∏–∞—Ç—É—Ä–æ–π
async def start_test(update: Update, context: CallbackContext):
    # –ü—Ä–µ–¥–ª–∞–≥–∞–µ–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é –≤—ã–±—Ä–∞—Ç—å —Ç–µ—Å—Ç
    keyboard = [
        [InlineKeyboardButton(test_name, callback_data=f"test_{i}")]  # –ò—Å–ø–æ–ª—å–∑—É–µ–º –∏–Ω–¥–µ–∫—Å –≤–º–µ—Å—Ç–æ –∏–º–µ–Ω–∏
        for i, test_name in enumerate(tests.keys())
    ]
    reply_markup = InlineKeyboardMarkup(keyboard)
    await update.message.reply_text("–í—ã–±–µ—Ä–∏—Ç–µ —Ç–µ—Å—Ç:", reply_markup=reply_markup)

# –û–±—Ä–∞–±–æ—Ç—á–∏–∫ –≤—ã–±–æ—Ä–∞ —Ç–µ—Å—Ç–∞
async def handle_test_selection(update: Update, context: CallbackContext):
    query = update.callback_query
    await query.answer()  # –û—Ç–≤–µ—Ç–∏–º –Ω–∞ callback, —á—Ç–æ–±—ã —É–±—Ä–∞—Ç—å "—á–∞—Å–∏–∫–∏" –Ω–∞ –∫–Ω–æ–ø–∫–µ

    test_index = int(query.data.replace("test_", ""))  # –ü–æ–ª—É—á–∞–µ–º –∏–Ω–¥–µ–∫—Å —Ç–µ—Å—Ç–∞
    test_name = list(tests.keys())[test_index]  # –ü–æ–ª—É—á–∞–µ–º –∏–º—è —Ç–µ—Å—Ç–∞ –ø–æ –∏–Ω–¥–µ–∫—Å—É

    if test_name in tests:
        context.user_data["test"] = {
            "name": test_name,
            "questions": tests[test_name],
            "current_question": 0,
            "score": 0
        }
        await ask_question(query, context)
    else:
        await query.edit_message_text("–¢–µ—Å—Ç –Ω–µ –Ω–∞–π–¥–µ–Ω. –ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –≤—ã–±–µ—Ä–∏—Ç–µ —Ç–µ—Å—Ç –∏–∑ —Å–ø–∏—Å–∫–∞.")

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –≤–æ–ø—Ä–æ—Å–∞ —Å –∏–Ω–ª–∞–π–Ω-–∫–ª–∞–≤–∏–∞—Ç—É—Ä–æ–π
async def ask_question(query, context: CallbackContext):
    user_data = context.user_data
    test = user_data.get("test")

    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –ª–∏ —Ç–µ—Å—Ç –≤ user_data
    if not test:
        await query.message.reply_text("–¢–µ—Å—Ç –Ω–µ –Ω–∞–π–¥–µ–Ω. –ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –Ω–∞—á–Ω–∏—Ç–µ –∑–∞–Ω–æ–≤–æ.")
        return

    current_question = test["current_question"]

    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å—Ç—å –ª–∏ –µ—â–µ –≤–æ–ø—Ä–æ—Å—ã
    if current_question < len(test["questions"]):
        question, options = test["questions"][current_question]
        keyboard = [
            [InlineKeyboardButton(option, callback_data=f"answer_{i}")]  # –ò—Å–ø–æ–ª—å–∑—É–µ–º –∏–Ω–¥–µ–∫—Å –≤–º–µ—Å—Ç–æ —Ç–µ–∫—Å—Ç–∞
            for i, option in enumerate(options.keys())
        ]
        reply_markup = InlineKeyboardMarkup(keyboard)
        await query.message.reply_text(question, reply_markup=reply_markup)  # –û—Ç–ø—Ä–∞–≤–ª—è–µ–º –Ω–æ–≤—ã–π –≤–æ–ø—Ä–æ—Å
    else:
        # –¢–µ—Å—Ç –∑–∞–≤–µ—Ä—à–µ–Ω, –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
        await finish_test(query, context)

# –û–±—Ä–∞–±–æ—Ç—á–∏–∫ –æ—Ç–≤–µ—Ç–∞ –Ω–∞ –≤–æ–ø—Ä–æ—Å
async def handle_test_answer(update: Update, context: CallbackContext):
    query = update.callback_query
    await query.answer()  # –û—Ç–≤–µ—Ç–∏–º –Ω–∞ callback, —á—Ç–æ–±—ã —É–±—Ä–∞—Ç—å "—á–∞—Å–∏–∫–∏" –Ω–∞ –∫–Ω–æ–ø–∫–µ

    user_data = context.user_data
    test = user_data.get("test")

    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –ª–∏ —Ç–µ—Å—Ç –≤ user_data
    if not test:
        await query.message.reply_text("–¢–µ—Å—Ç –Ω–µ –Ω–∞–π–¥–µ–Ω. –ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –Ω–∞—á–Ω–∏—Ç–µ –∑–∞–Ω–æ–≤–æ.")
        return

    current_question = test["current_question"]
    answer_index = int(query.data.replace("answer_", ""))  # –ü–æ–ª—É—á–∞–µ–º –∏–Ω–¥–µ–∫—Å –æ—Ç–≤–µ—Ç–∞

    # –ù–∞—Ö–æ–¥–∏–º –≤—ã–±—Ä–∞–Ω–Ω—ã–π –≤–∞—Ä–∏–∞–Ω—Ç –∏ –¥–æ–±–∞–≤–ª—è–µ–º –±–∞–ª–ª—ã
    question, options = test["questions"][current_question]
    option_text = list(options.keys())[answer_index]  # –ü–æ–ª—É—á–∞–µ–º —Ç–µ–∫—Å—Ç –æ—Ç–≤–µ—Ç–∞ –ø–æ –∏–Ω–¥–µ–∫—Å—É
    if option_text in options:
        test["score"] += options[option_text]

    # –£–¥–∞–ª—è–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ —Å –≤–æ–ø—Ä–æ—Å–æ–º
    try:
        await query.delete_message()
    except Exception as e:
        logger.error(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ —É–¥–∞–ª–µ–Ω–∏–∏ —Å–æ–æ–±—â–µ–Ω–∏—è: {e}")

    # –ü–µ—Ä–µ—Ö–æ–¥–∏–º –∫ —Å–ª–µ–¥—É—é—â–µ–º—É –≤–æ–ø—Ä–æ—Å—É
    test["current_question"] += 1

    # –û—Ç–ø—Ä–∞–≤–ª—è–µ–º —Å–ª–µ–¥—É—é—â–∏–π –≤–æ–ø—Ä–æ—Å
    await ask_question(query, context)

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∑–∞–≤–µ—Ä—à–µ–Ω–∏—è —Ç–µ—Å—Ç–∞
async def finish_test(query, context: CallbackContext):
    user_data = context.user_data
    test = user_data["test"]
    total_score = test["score"]

    # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä –ø–µ—á–∞—Ç–∏
    await query.message.chat.send_action(action=ChatAction.TYPING)

    # –§–æ—Ä–º–∏—Ä—É–µ–º –ø—Ä–æ–º—Ç –¥–ª—è –º–æ–¥–µ–ª–∏
    prompt = (
        f"–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –ø—Ä–æ—à–µ–ª —Ç–µ—Å—Ç '{test['name']}' –∏ –Ω–∞–±—Ä–∞–ª {total_score} –±–∞–ª–ª–æ–≤. "
        f"–°–≥–µ–Ω–µ—Ä–∏—Ä—É–π –ø–æ–¥—Ä–æ–±–Ω—ã–π –∞–Ω–∞–ª–∏–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∏ –¥–∞–π —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏."
    )

    # –û–ø—Ä–µ–¥–µ–ª—è–µ–º callback –¥–ª—è –æ—Ç–ø—Ä–∞–≤–∫–∏ –æ—Ç–≤–µ—Ç–∞
    async def send_response(response):
        await query.message.reply_text(response, reply_markup=get_keyboard())

    # –î–æ–±–∞–≤–ª—è–µ–º –∑–∞–ø—Ä–æ—Å –≤ –æ—á–µ—Ä–µ–¥—å
    user_id = query.from_user.id
    message_timestamps[user_id] = time.time()
    message_queue.put({
        "user_id": user_id,
        "question": prompt,
        "callback": send_response,
        "user_name": user_data.get("user_name")
    })

    user_data["state"] = None  # –°–±—Ä–∞—Å—ã–≤–∞–µ–º —Å–æ—Å—Ç–æ—è–Ω–∏–µ



def is_complex_query(text):
    complex_keywords = [
        "—Å–æ—Å—Ç–∞–≤—å –ø–ª–∞–Ω", "–ø–ª–∞–Ω —Å–∞–º–æ—Ä–∞–∑–≤–∏—Ç–∏—è", "–∞–Ω–∞–ª–∏–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤", "–ø–æ–º–æ–≥–∏ —Å —Ü–µ–ª—è–º–∏"
    ]
    return any(keyword in text.lower() for keyword in complex_keywords)

# –û–±—Ä–∞–±–æ—Ç—á–∏–∫ —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö —Å–æ–æ–±—â–µ–Ω–∏–π
async def create_self_improvement_plan(update: Update, context: CallbackContext):
    user_data = context.user_data

    # –®–∞–≥ 1: –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å —Ü–µ–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
    await update.message.reply_text("–î–∞–≤–∞–π—Ç–µ –Ω–∞—á–Ω–µ–º —Å –≤–∞—à–∏—Ö —Ü–µ–ª–µ–π. –ö–∞–∫–∏–µ –æ–±–ª–∞—Å—Ç–∏ –≤—ã —Ö–æ—Ç–∏—Ç–µ —É–ª—É—á—à–∏—Ç—å? (–Ω–∞–ø—Ä–∏–º–µ—Ä, –∫–∞—Ä—å–µ—Ä–∞, –∑–¥–æ—Ä–æ–≤—å–µ, –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ)")
    user_data["state"] = "waiting_for_goals"






# –û–±—Ä–∞–±–æ—Ç—á–∏–∫ –≥–æ–ª–æ—Å–æ–≤—ã—Ö —Å–æ–æ–±—â–µ–Ω–∏–π
async def handle_voice(update: Update, context: CallbackContext):
    user_data = context.user_data

    # –°–Ω–∞—á–∞–ª–∞ –ø–æ–∫–∞–∑—ã–≤–∞–µ–º, —á—Ç–æ –±–æ—Ç —Å–ª—É—à–∞–µ—Ç
    await update.message.chat.send_action(action=ChatAction.RECORD_VOICE)

    voice_file = await update.message.voice.get_file()
    ogg_path = "voice_message.ogg"
    await voice_file.download_to_drive(ogg_path)
    text = transcribe_audio(ogg_path)  # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –≥–æ–ª–æ—Å–∞ –≤ —Ç–µ–∫—Å—Ç

    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ "–ø–æ–Ω—è—Ç–Ω–æ—Å—Ç—å" –∑–∞–ø—Ä–æ—Å–∞
    if not is_message_clear(text):
        await update.message.reply_text(
            "–ò–∑–≤–∏–Ω–∏—Ç–µ, —è –Ω–µ —Å–æ–≤—Å–µ–º –ø–æ–Ω—è–ª –≤–∞—à –∑–∞–ø—Ä–æ—Å. –ü–æ–∂–∞–ª—É–π—Å—Ç–∞, —É—Ç–æ—á–Ω–∏—Ç–µ, —á—Ç–æ –≤—ã –∏–º–µ–µ—Ç–µ –≤ –≤–∏–¥—É?\n\n"
            "–ù–∞–ø—Ä–∏–º–µ—Ä, –≤—ã –º–æ–∂–µ—Ç–µ —Å–ø—Ä–æ—Å–∏—Ç—å:\n"
            "- –ö–∞–∫ –ø–æ–¥–≥–æ—Ç–æ–≤–∏—Ç—å—Å—è –∫ —ç–∫–∑–∞–º–µ–Ω–∞–º?\n"
            "- –ß—Ç–æ –¥–µ–ª–∞—Ç—å –ø—Ä–∏ —Ñ–∏–Ω–∞–Ω—Å–æ–≤–æ–π –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏?\n"
            "- –ö–∞–∫ –Ω–∞–π—Ç–∏ —Å–≤–æ–µ –ø—Ä–∏–∑–≤–∞–Ω–∏–µ?"
        )
        return

    # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä –ø–µ—á–∞—Ç–∏
    await send_typing_action(update)

    # –û–ø—Ä–µ–¥–µ–ª—è–µ–º callback –¥–ª—è –æ—Ç–ø—Ä–∞–≤–∫–∏ –æ—Ç–≤–µ—Ç–∞
    async def send_voice_response(response):
        await update.message.reply_text(response)
        await text_to_speech(update, response)

    # –î–æ–±–∞–≤–ª—è–µ–º –∑–∞–ø—Ä–æ—Å –≤ –æ—á–µ—Ä–µ–¥—å
    user_id = update.effective_user.id
    user_name = user_data.get("user_name")
    message_timestamps[user_id] = time.time()
    message_queue.put({
        "user_id": user_id,
        "question": text,
        "callback": send_voice_response,
        "user_name": user_name
    })

from telegram.ext import ConversationHandler

# –û–ø—Ä–µ–¥–µ–ª—è–µ–º —Å–æ—Å—Ç–æ—è–Ω–∏—è –¥–∏–∞–ª–æ–≥–∞
WAITING_FOR_GOALS, WAITING_FOR_RESOURCES = range(2)



# –û–±—Ä–∞–±–æ—Ç—á–∏–∫ –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —Ü–µ–ª–µ–π
async def handle_goals(update: Update, context: CallbackContext):
    goals = update.message.text
    context.user_data["goals"] = goals

    # –ü–µ—Ä–µ—Ö–æ–¥–∏–º –∫ —Å–ª–µ–¥—É—é—â–µ–º—É —à–∞–≥—É
    await update.message.reply_text("–û—Ç–ª–∏—á–Ω–æ! –¢–µ–ø–µ—Ä—å —É–∫–∞–∂–∏—Ç–µ, –∫–∞–∫–∏–µ —Ä–µ—Å—É—Ä—Å—ã —É –≤–∞—Å —É–∂–µ –µ—Å—Ç—å (–Ω–∞–ø—Ä–∏–º–µ—Ä, –∫–Ω–∏–≥–∏, –∫—É—Ä—Å—ã, –Ω–∞—Å—Ç–∞–≤–Ω–∏–∫–∏).")
    return WAITING_FOR_RESOURCES  # –ü–µ—Ä–µ—Ö–æ–¥–∏–º –∫ —Å–æ—Å—Ç–æ—è–Ω–∏—é –æ–∂–∏–¥–∞–Ω–∏—è —Ä–µ—Å—É—Ä—Å–æ–≤

# –û–±—Ä–∞–±–æ—Ç—á–∏–∫ –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —Ä–µ—Å—É—Ä—Å–æ–≤
async def handle_resources(update: Update, context: CallbackContext):
    resources = update.message.text
    context.user_data["resources"] = resources

    # –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –ø–ª–∞–Ω–∞ —Å –ø–æ–º–æ—â—å—é –º–æ–¥–µ–ª–∏
    goals = context.user_data.get("goals", "–Ω–µ —É–∫–∞–∑–∞–Ω—ã")
    resources = context.user_data.get("resources", "–Ω–µ —É–∫–∞–∑–∞–Ω—ã")

    # –§–æ—Ä–º–∏—Ä—É–µ–º –ø—Ä–æ–º—Ç –¥–ª—è –º–æ–¥–µ–ª–∏
    prompt = (
        f"–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å —Ö–æ—á–µ—Ç —É–ª—É—á—à–∏—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ –æ–±–ª–∞—Å—Ç–∏: {goals}. "
        f"–£ –Ω–µ–≥–æ –µ—Å—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ —Ä–µ—Å—É—Ä—Å—ã: {resources}. "
        f"–°–æ—Å—Ç–∞–≤—å –ø–æ–¥—Ä–æ–±–Ω—ã–π –ø–ª–∞–Ω —Å–∞–º–æ—Ä–∞–∑–≤–∏—Ç–∏—è, –≤–∫–ª—é—á–∞—é—â–∏–π –∫–æ–Ω–∫—Ä–µ—Ç–Ω—ã–µ —à–∞–≥–∏ –∏ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏."
    )

    # –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –ø–ª–∞–Ω–∞ —Å –ø–æ–º–æ—â—å—é –º–æ–¥–µ–ª–∏
    plan = generate_response(prompt)

    # –û—Ç–ø—Ä–∞–≤–∫–∞ –ø–ª–∞–Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é
    await update.message.reply_text(plan, reply_markup=get_keyboard())

    # –ó–∞–≤–µ—Ä—à–∞–µ–º –¥–∏–∞–ª–æ–≥
    return ConversationHandler.END

# –û–±—Ä–∞–±–æ—Ç—á–∏–∫ –æ—Ç–º–µ–Ω—ã –¥–∏–∞–ª–æ–≥–∞
async def cancel(update: Update, context: CallbackContext):
    await update.message.reply_text("–î–∏–∞–ª–æ–≥ –æ—Ç–º–µ–Ω–µ–Ω.", reply_markup=get_keyboard())
    return ConversationHandler.END

import sqlite3
from telegram.ext import ConversationHandler

# –û–ø—Ä–µ–¥–µ–ª—è–µ–º —Å–æ—Å—Ç–æ—è–Ω–∏—è –¥–∏–∞–ª–æ–≥–∞ –¥–ª—è —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏
WAITING_FOR_NAME, WAITING_FOR_AGE, WAITING_FOR_INTERESTS, WAITING_FOR_PROBLEMS = range(4)

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö
def init_db():
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS users (
            user_id INTEGER PRIMARY KEY,
            username TEXT,
            first_name TEXT,
            last_name TEXT,
            age INTEGER,
            interests TEXT,
            problems TEXT,
            registered_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
        )
    ''')
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS goals (
            goal_id INTEGER PRIMARY KEY AUTOINCREMENT,
            user_id INTEGER,
            description TEXT,
            deadline TEXT,
            progress INTEGER DEFAULT 0,
            status TEXT DEFAULT '–í –ø—Ä–æ—Ü–µ—Å—Å–µ',
            FOREIGN KEY (user_id) REFERENCES users (user_id)
        )
    ''')
    conn.commit()
    conn.close()

# –†–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
def register_user(user_id, username, first_name, last_name, age, interests, problems):
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    cursor.execute('''
        INSERT OR REPLACE INTO users (user_id, username, first_name, last_name, age, interests, problems)
        VALUES (?, ?, ?, ?, ?, ?, ?)
    ''', (user_id, username, first_name, last_name, age, interests, problems))
    conn.commit()
    conn.close()

# –ü–æ–ª—É—á–µ–Ω–∏–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ
def get_user_info(user_id):
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    cursor.execute('SELECT * FROM users WHERE user_id = ?', (user_id,))
    user_info = cursor.fetchone()
    conn.close()
    return user_info

# –û–±—Ä–∞–±–æ—Ç—á–∏–∫ –∫–æ–º–∞–Ω–¥—ã /start
async def start(update: Update, context: CallbackContext):
    user = update.effective_user
    user_info = get_user_info(user.id)

    # –ï—Å–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å —É–∂–µ –∑–∞—Ä–µ–≥–∏—Å—Ç—Ä–∏—Ä–æ–≤–∞–Ω
    if user_info:
        await update.message.reply_text(
            f"üëã –ü—Ä–∏–≤–µ—Ç, {user_info[2]}! –†–∞–¥ —Å–Ω–æ–≤–∞ —Ç–µ–±—è –≤–∏–¥–µ—Ç—å. –ß–µ–º –º–æ–≥—É –ø–æ–º–æ—á—å?",
            reply_markup=get_keyboard()
        )
        return ConversationHandler.END

    # –ï—Å–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –Ω–µ –∑–∞—Ä–µ–≥–∏—Å—Ç—Ä–∏—Ä–æ–≤–∞–Ω
    await update.message.reply_text("üëã –ü—Ä–∏–≤–µ—Ç! –î–∞–≤–∞–π –ø–æ–∑–Ω–∞–∫–æ–º–∏–º—Å—è. –ö–∞–∫ —Ç–µ–±—è –∑–æ–≤—É—Ç?")
    return WAITING_FOR_NAME

# –û–±—Ä–∞–±–æ—Ç—á–∏–∫ –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –∏–º–µ–Ω–∏
async def handle_name(update: Update, context: CallbackContext):
    user = update.effective_user
    context.user_data["first_name"] = update.message.text
    await update.message.reply_text("–ü—Ä–∏—è—Ç–Ω–æ –ø–æ–∑–Ω–∞–∫–æ–º–∏—Ç—å—Å—è! –°–∫–æ–ª—å–∫–æ —Ç–µ–±–µ –ª–µ—Ç?")
    return WAITING_FOR_AGE

# –û–±—Ä–∞–±–æ—Ç—á–∏–∫ –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –≤–æ–∑—Ä–∞—Å—Ç–∞
async def handle_age(update: Update, context: CallbackContext):
    try:
        age = int(update.message.text)
        if age < 0 or age > 120:
            await update.message.reply_text("–ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –≤–≤–µ–¥–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–æ–∑—Ä–∞—Å—Ç (–æ—Ç 0 –¥–æ 120).")
            return WAITING_FOR_AGE
        context.user_data["age"] = age
        await update.message.reply_text("–û—Ç–ª–∏—á–Ω–æ! –ö–∞–∫–∏–µ —É —Ç–µ–±—è –∏–Ω—Ç–µ—Ä–µ—Å—ã? (–Ω–∞–ø—Ä–∏–º–µ—Ä, —Å–ø–æ—Ä—Ç, –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏–µ, –º—É–∑—ã–∫–∞)")
        return WAITING_FOR_INTERESTS
    except ValueError:
        await update.message.reply_text("–ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –≤–≤–µ–¥–∏—Ç–µ –≤–æ–∑—Ä–∞—Å—Ç —á–∏—Å–ª–æ–º.")
        return WAITING_FOR_AGE

# –û–±—Ä–∞–±–æ—Ç—á–∏–∫ –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –∏–Ω—Ç–µ—Ä–µ—Å–æ–≤
async def handle_interests(update: Update, context: CallbackContext):
    context.user_data["interests"] = update.message.text
    await update.message.reply_text("–•–æ—Ä–æ—à–æ! –ö–∞–∫–∏–µ —É —Ç–µ–±—è –æ—Å–Ω–æ–≤–Ω—ã–µ –ø—Ä–æ–±–ª–µ–º—ã –∏–ª–∏ –≤–æ–ø—Ä–æ—Å—ã, —Å –∫–æ—Ç–æ—Ä—ã–º–∏ —Ç—ã —Ö–æ—Ç–µ–ª –±—ã –ø–æ–º–æ—á—å?")
    return WAITING_FOR_PROBLEMS

# –û–±—Ä–∞–±–æ—Ç—á–∏–∫ –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –ø—Ä–æ–±–ª–µ–º
async def handle_problems(update: Update, context: CallbackContext):
    user = update.effective_user
    problems = update.message.text
    context.user_data["problems"] = problems

    # –°–æ—Ö—Ä–∞–Ω—è–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤ –±–∞–∑—É –¥–∞–Ω–Ω—ã—Ö
    register_user(
        user.id,
        user.username,
        context.user_data.get("first_name"),
        user.last_name,
        context.user_data.get("age"),
        context.user_data.get("interests"),
        problems
    )

    await update.message.reply_text(
        "–°–ø–∞—Å–∏–±–æ –∑–∞ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏—é! –¢–µ–ø–µ—Ä—å —è –∑–Ω–∞—é –æ —Ç–µ–±–µ –Ω–µ–º–Ω–æ–≥–æ –±–æ–ª—å—à–µ. –ß–µ–º –º–æ–≥—É –ø–æ–º–æ—á—å?",
        reply_markup=get_keyboard()
    )
    return ConversationHandler.END

# –û–±—Ä–∞–±–æ—Ç—á–∏–∫ –æ—Ç–º–µ–Ω—ã –¥–∏–∞–ª–æ–≥–∞
async def cancel(update: Update, context: CallbackContext):
    await update.message.reply_text("–†–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏—è –æ—Ç–º–µ–Ω–µ–Ω–∞.", reply_markup=get_keyboard())
    return ConversationHandler.END

# –ö–ª–∞–≤–∏–∞—Ç—É—Ä–∞ —Å –∫–Ω–æ–ø–∫–∞–º–∏
def get_keyboard():
    keyboard = [
        ["–ó–∞–¥–∞—Ç—å –≤–æ–ø—Ä–æ—Åü§î", "–ü—Ä–æ—Ñ–∏–ª—å"],
        ["–ù–∞—à –∫–∞–Ω–∞–ª!", "–ü–æ–¥–¥–µ—Ä–∂–∫–∞üòå", "–¢–µ—Å—Ç"]
    ]
    return ReplyKeyboardMarkup(keyboard, resize_keyboard=True)

# –û–±—Ä–∞–±–æ—Ç—á–∏–∫ –¥–ª—è –∫–Ω–æ–ø–∫–∏ "–ü—Ä–æ—Ñ–∏–ª—å"
async def show_profile(update: Update, context: CallbackContext):
    user = update.effective_user
    user_info = get_user_info(user.id)

    # –ï—Å–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –∑–∞—Ä–µ–≥–∏—Å—Ç—Ä–∏—Ä–æ–≤–∞–Ω
    if user_info:
        profile_message = (
            f"üë§ –í–∞—à –ø—Ä–æ—Ñ–∏–ª—å:\n\n"
            f"–ò–º—è: {user_info[2]}\n"
            f"–í–æ–∑—Ä–∞—Å—Ç: {user_info[4]}\n"
            f"–ò–Ω—Ç–µ—Ä–µ—Å—ã: {user_info[5]}\n"
            f"–û—Å–Ω–æ–≤–Ω—ã–µ –ø—Ä–æ–±–ª–µ–º—ã: {user_info[6]}"
        )
        await update.message.reply_text(profile_message, reply_markup=get_keyboard())
    else:
        await update.message.reply_text("–í—ã –µ—â–µ –Ω–µ –∑–∞—Ä–µ–≥–∏—Å—Ç—Ä–∏—Ä–æ–≤–∞–Ω—ã. –ù–∞–∂–º–∏—Ç–µ /start, —á—Ç–æ–±—ã –Ω–∞—á–∞—Ç—å.", reply_markup=get_keyboard())

async def handle_message(update: Update, context: CallbackContext):
    user_message = update.message.text
    user_data = context.user_data

    # –ï—Å–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ —Å–æ—Å—Ç–æ—è–Ω–∏–∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏—è —Ü–µ–ª–∏
    if user_data.get("state") in ["WAITING_FOR_GOAL_DESCRIPTION", "WAITING_FOR_GOAL_DEADLINE"]:
        # –ü–µ—Ä–µ–¥–∞–µ–º —É–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –æ–±—Ä–∞–±–æ—Ç—á–∏–∫–∞–º —Ü–µ–ª–µ–π
        return await handle_goal_description(update, context) if user_data.get("state") == "WAITING_FOR_GOAL_DESCRIPTION" else await handle_goal_deadline(update, context)

    # –ï—Å–ª–∏ –±–æ—Ç –∂–¥—ë—Ç –∏–º—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
    if user_data.get("state") == "waiting_for_name":
        user_data["user_name"] = user_message  # –°–æ—Ö—Ä–∞–Ω—è–µ–º –∏–º—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
        user_data["state"] = None  # –°–±—Ä–∞—Å—ã–≤–∞–µ–º —Å–æ—Å—Ç–æ—è–Ω–∏–µ
        response = f"–ü—Ä–∏—è—Ç–Ω–æ –ø–æ–∑–Ω–∞–∫–æ–º–∏—Ç—å—Å—è, {user_message}! –ß–µ–º –º–æ–≥—É –ø–æ–º–æ—á—å?"
        await update.message.reply_text(response)
        await text_to_speech(update, response)  # –û—Ç–ø—Ä–∞–≤–∫–∞ –≥–æ–ª–æ—Å–æ–≤–æ–≥–æ –æ—Ç–≤–µ—Ç–∞
        return

    # –û–±—Ä–∞–±–æ—Ç–∫–∞ –∫–Ω–æ–ø–æ–∫
    if user_message == "–¢–µ—Å—Ç":
        await start_test(update, context)
    elif user_message == "–ó–∞–¥–∞—Ç—å –≤–æ–ø—Ä–æ—Åü§î":
        response = "–ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –∑–∞–¥–∞–π—Ç–µ –≤–∞—à –≤–æ–ø—Ä–æ—Å —Ç–µ–∫—Å—Ç–æ–º –∏–ª–∏ –æ—Ç–ø—Ä–∞–≤—å—Ç–µ –≥–æ–ª–æ—Å–æ–≤–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ."
        await update.message.reply_text(response)
        await text_to_speech(update, response)  # –û—Ç–ø—Ä–∞–≤–∫–∞ –≥–æ–ª–æ—Å–æ–≤–æ–≥–æ –æ—Ç–≤–µ—Ç–∞
    elif user_message == "–ù–∞—à –∫–∞–Ω–∞–ª!":
        response = "–ü–æ–¥–ø–∏—Å—ã–≤–∞–π—Ç–µ—Å—å –Ω–∞ –Ω–∞—à –∫–∞–Ω–∞–ª: https://t.me/mentachannel"
        await update.message.reply_text(response)
        await text_to_speech(update, response)  # –û—Ç–ø—Ä–∞–≤–∫–∞ –≥–æ–ª–æ—Å–æ–≤–æ–≥–æ –æ—Ç–≤–µ—Ç–∞
    elif user_message == "–ü–æ–¥–¥–µ—Ä–∂–∫–∞üòå":
        response = "–°–≤—è–∂–∏—Ç–µ—Å—å —Å –ø–æ–¥–¥–µ—Ä–∂–∫–æ–π: @adinaadinaadina"
        await update.message.reply_text(response)
        await text_to_speech(update, response)  # –û—Ç–ø—Ä–∞–≤–∫–∞ –≥–æ–ª–æ—Å–æ–≤–æ–≥–æ –æ—Ç–≤–µ—Ç–∞
    elif user_message == "–ü—Ä–æ—Ñ–∏–ª—å":
        await show_profile(update, context)
    elif user_message == "–î–æ–±–∞–≤–∏—Ç—å —Ü–µ–ª—å":  # –û–±—Ä–∞–±–æ—Ç–∫–∞ –∫–Ω–æ–ø–∫–∏ "–î–æ–±–∞–≤–∏—Ç—å —Ü–µ–ª—å"
        await add_goal(update, context)
    elif user_message == "–ú–æ–∏ —Ü–µ–ª–∏":  # –û–±—Ä–∞–±–æ—Ç–∫–∞ –∫–Ω–æ–ø–∫–∏ "–ú–æ–∏ —Ü–µ–ª–∏"
        await show_goals(update, context)
    elif user_message == "–û—Ç–º–µ—Ç–∏—Ç—å –ø—Ä–æ–≥—Ä–µ—Å—Å":  # –û–±—Ä–∞–±–æ—Ç–∫–∞ –∫–Ω–æ–ø–∫–∏ "–û—Ç–º–µ—Ç–∏—Ç—å –ø—Ä–æ–≥—Ä–µ—Å—Å"
        await update_progress(update, context)
    elif user_message == "–£–¥–∞–ª–∏—Ç—å —Ü–µ–ª—å":  # –û–±—Ä–∞–±–æ—Ç–∫–∞ –∫–Ω–æ–ø–∫–∏ "–£–¥–∞–ª–∏—Ç—å —Ü–µ–ª—å"
        await delete_goal(update, context)
    else:
        # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ "–ø–æ–Ω—è—Ç–Ω–æ—Å—Ç—å" –∑–∞–ø—Ä–æ—Å–∞
        if not is_message_clear(user_message):
            response = (
                "–ò–∑–≤–∏–Ω–∏—Ç–µ, —è –Ω–µ —Å–æ–≤—Å–µ–º –ø–æ–Ω—è–ª –≤–∞—à –∑–∞–ø—Ä–æ—Å. –ü–æ–∂–∞–ª—É–π—Å—Ç–∞, —É—Ç–æ—á–Ω–∏—Ç–µ, —á—Ç–æ –≤—ã –∏–º–µ–µ—Ç–µ –≤ –≤–∏–¥—É?\n\n"
                "–ù–∞–ø—Ä–∏–º–µ—Ä, –≤—ã –º–æ–∂–µ—Ç–µ —Å–ø—Ä–æ—Å–∏—Ç—å:\n"
                "- –ö–∞–∫ –ø–æ–¥–≥–æ—Ç–æ–≤–∏—Ç—å—Å—è –∫ —ç–∫–∑–∞–º–µ–Ω–∞–º?\n"
                "- –ß—Ç–æ –¥–µ–ª–∞—Ç—å –ø—Ä–∏ —Ñ–∏–Ω–∞–Ω—Å–æ–≤–æ–π –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏?\n"
                "- –ö–∞–∫ –Ω–∞–π—Ç–∏ —Å–≤–æ–µ –ø—Ä–∏–∑–≤–∞–Ω–∏–µ?"
            )
            await update.message.reply_text(response)
            await text_to_speech(update, response)  # –û—Ç–ø—Ä–∞–≤–∫–∞ –≥–æ–ª–æ—Å–æ–≤–æ–≥–æ –æ—Ç–≤–µ—Ç–∞
            return

        # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä –ø–µ—á–∞—Ç–∏
        await send_typing_action(update)

        # –û–ø—Ä–µ–¥–µ–ª—è–µ–º callback –¥–ª—è –æ—Ç–ø—Ä–∞–≤–∫–∏ –æ—Ç–≤–µ—Ç–∞
        async def send_text_response(response):
            await update.message.reply_text(response)
            await text_to_speech(update, response)  # –û—Ç–ø—Ä–∞–≤–∫–∞ –≥–æ–ª–æ—Å–æ–≤–æ–≥–æ –æ—Ç–≤–µ—Ç–∞

        # –î–æ–±–∞–≤–ª—è–µ–º –∑–∞–ø—Ä–æ—Å –≤ –æ—á–µ—Ä–µ–¥—å
        user_id = update.effective_user.id
        user_name = user_data.get("user_name")
        message_timestamps[user_id] = time.time()
        message_queue.put({
            "user_id": user_id,
            "question": user_message,
            "callback": send_text_response,
            "user_name": user_name
        })

from telegram import ReplyKeyboardMarkup

# –ö–ª–∞–≤–∏–∞—Ç—É—Ä–∞ –¥–ª—è —É–ø—Ä–∞–≤–ª–µ–Ω–∏—è —Ü–µ–ª—è–º–∏
def get_goals_keyboard():
    keyboard = [
        ["–î–æ–±–∞–≤–∏—Ç—å —Ü–µ–ª—å", "–ú–æ–∏ —Ü–µ–ª–∏"],
        ["–û—Ç–º–µ—Ç–∏—Ç—å –ø—Ä–æ–≥—Ä–µ—Å—Å", "–£–¥–∞–ª–∏—Ç—å —Ü–µ–ª—å"],
        ["–ù–∞–∑–∞–¥"]
    ]
    return ReplyKeyboardMarkup(keyboard, resize_keyboard=True)

# –û–±—Ä–∞–±–æ—Ç—á–∏–∫ –∫–æ–º–∞–Ω–¥—ã /goals
async def goals(update: Update, context: CallbackContext):
    await update.message.reply_text(
        "üéØ –£–ø—Ä–∞–≤–ª–µ–Ω–∏–µ —Ü–µ–ª—è–º–∏. –í—ã–±–µ—Ä–∏—Ç–µ –¥–µ–π—Å—Ç–≤–∏–µ:",
        reply_markup=get_goals_keyboard()
    )


async def add_goal(update: Update, context: CallbackContext):
    await update.message.reply_text("–ù–∞–ø–∏—à–∏—Ç–µ –æ–ø–∏—Å–∞–Ω–∏–µ –≤–∞—à–µ–π —Ü–µ–ª–∏:")
    context.user_data["state"] = "WAITING_FOR_GOAL_DESCRIPTION"  # –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º —Å–æ—Å—Ç–æ—è–Ω–∏–µ
    return "WAITING_FOR_GOAL_DESCRIPTION"

async def handle_goal_description(update: Update, context: CallbackContext):
    description = update.message.text
    context.user_data["goal_description"] = description
    await update.message.reply_text("–£–∫–∞–∂–∏—Ç–µ —Å—Ä–æ–∫ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —Ü–µ–ª–∏ (–Ω–∞–ø—Ä–∏–º–µ—Ä, 2023-12-31):")
    context.user_data["state"] = "WAITING_FOR_GOAL_DEADLINE"  # –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º —Å–æ—Å—Ç–æ—è–Ω–∏–µ
    return "WAITING_FOR_GOAL_DEADLINE"

async def handle_goal_deadline(update: Update, context: CallbackContext):
    deadline = update.message.text
    user_id = update.effective_user.id
    description = context.user_data["goal_description"]

    # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ü–µ–ª—å –≤ –±–∞–∑—É –¥–∞–Ω–Ω—ã—Ö
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    cursor.execute('''
        INSERT INTO goals (user_id, description, deadline)
        VALUES (?, ?, ?)
    ''', (user_id, description, deadline))
    conn.commit()
    conn.close()

    await update.message.reply_text("–¶–µ–ª—å —É—Å–ø–µ—à–Ω–æ –¥–æ–±–∞–≤–ª–µ–Ω–∞!", reply_markup=get_goals_keyboard())
    context.user_data["state"] = None  # –°–±—Ä–∞—Å—ã–≤–∞–µ–º —Å–æ—Å—Ç–æ—è–Ω–∏–µ
    return ConversationHandler.END


async def show_goals(update: Update, context: CallbackContext):
    user_id = update.effective_user.id
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    cursor.execute('SELECT description, deadline, progress, status FROM goals WHERE user_id = ?', (user_id,))
    goals = cursor.fetchall()
    conn.close()

    if not goals:
        await update.message.reply_text("–£ –≤–∞—Å –ø–æ–∫–∞ –Ω–µ—Ç —Ü–µ–ª–µ–π.", reply_markup=get_goals_keyboard())
        return

    goals_message = "üéØ –í–∞—à–∏ —Ü–µ–ª–∏:\n\n"
    for goal in goals:
        goals_message += f"‚Ä¢ {goal[0]}\n–°—Ä–æ–∫: {goal[1]}\n–ü—Ä–æ–≥—Ä–µ—Å—Å: {goal[2]}%\n–°—Ç–∞—Ç—É—Å: {goal[3]}\n\n"

    await update.message.reply_text(goals_message, reply_markup=get_goals_keyboard())


async def update_progress(update: Update, context: CallbackContext):
    user_id = update.effective_user.id
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    cursor.execute('SELECT goal_id, description FROM goals WHERE user_id = ?', (user_id,))
    goals = cursor.fetchall()
    conn.close()

    if not goals:
        await update.message.reply_text("–£ –≤–∞—Å –ø–æ–∫–∞ –Ω–µ—Ç —Ü–µ–ª–µ–π.", reply_markup=get_goals_keyboard())
        return

    keyboard = [
        [InlineKeyboardButton(goal[1], callback_data=f"progress_{goal[0]}")]
        for goal in goals
    ]
    reply_markup = InlineKeyboardMarkup(keyboard)
    await update.message.reply_text("–í—ã–±–µ—Ä–∏—Ç–µ —Ü–µ–ª—å –¥–ª—è –æ–±–Ω–æ–≤–ª–µ–Ω–∏—è –ø—Ä–æ–≥—Ä–µ—Å—Å–∞:", reply_markup=reply_markup)

async def check_goal_completion(update: Update, context: CallbackContext):
    user_id = update.effective_user.id
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    cursor.execute('SELECT goal_id, description FROM goals WHERE user_id = ?', (user_id,))
    goals = cursor.fetchall()
    conn.close()

    if not goals:
        await update.message.reply_text("–£ –≤–∞—Å –ø–æ–∫–∞ –Ω–µ—Ç —Ü–µ–ª–µ–π.", reply_markup=get_goals_keyboard())
        return

    # –°–æ–∑–¥–∞–µ–º –∏–Ω–ª–∞–π–Ω-–∫–Ω–æ–ø–∫–∏ –¥–ª—è –∫–∞–∂–¥–æ–π —Ü–µ–ª–∏
    keyboard = [
        [InlineKeyboardButton(goal[1], callback_data=f"complete_{goal[0]}")]
        for goal in goals
    ]
    reply_markup = InlineKeyboardMarkup(keyboard)
    await update.message.reply_text("–í—ã–±–µ—Ä–∏—Ç–µ —Ü–µ–ª—å –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è:", reply_markup=reply_markup)


async def handle_goal_selection(update: Update, context: CallbackContext):
    query = update.callback_query
    await query.answer()

    # –ò–∑–≤–ª–µ–∫–∞–µ–º goal_id –∏–∑ callback_data
    goal_id = int(query.data.replace("complete_", ""))

    # –°–æ—Ö—Ä–∞–Ω—è–µ–º goal_id –≤ user_data –¥–ª—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –≤ —Å–ª–µ–¥—É—é—â–µ–º —à–∞–≥–µ
    context.user_data["goal_id"] = goal_id

    # –ó–∞–ø—Ä–∞—à–∏–≤–∞–µ–º, –≤—ã–ø–æ–ª–Ω–µ–Ω–∞ –ª–∏ —Ü–µ–ª—å
    await query.edit_message_text("–¶–µ–ª—å –≤—ã–ø–æ–ª–Ω–µ–Ω–∞? (–î–∞/–ù–µ—Ç):")
    return "WAITING_FOR_COMPLETION_STATUS"


async def handle_completion_status(update: Update, context: CallbackContext):
    user_response = update.message.text.lower()

    # –ü—Ä–æ–≤–µ—Ä—è–µ–º –æ—Ç–≤–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
    if user_response not in ["–¥–∞", "–Ω–µ—Ç"]:
        await update.message.reply_text("–ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –æ—Ç–≤–µ—Ç—å—Ç–µ '–î–∞' –∏–ª–∏ '–ù–µ—Ç'.")
        return "WAITING_FOR_COMPLETION_STATUS"

    # –ò–∑–≤–ª–µ–∫–∞–µ–º goal_id –∏–∑ user_data
    goal_id = context.user_data.get("goal_id")
    if not goal_id:
        await update.message.reply_text("–û—à–∏–±–∫–∞: —Ü–µ–ª—å –Ω–µ –Ω–∞–π–¥–µ–Ω–∞. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ —Å–Ω–æ–≤–∞.")
        return ConversationHandler.END

    # –û–±–Ω–æ–≤–ª—è–µ–º —Å—Ç–∞—Ç—É—Å —Ü–µ–ª–∏ –≤ –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    if user_response == "–¥–∞":
        cursor.execute('UPDATE goals SET status = ? WHERE goal_id = ?', ("–ó–∞–≤–µ—Ä—à–µ–Ω–æ", goal_id))
        await update.message.reply_text("–¶–µ–ª—å –æ—Ç–º–µ—á–µ–Ω–∞ –∫–∞–∫ –≤—ã–ø–æ–ª–Ω–µ–Ω–Ω–∞—è!", reply_markup=get_goals_keyboard())
    else:
        cursor.execute('UPDATE goals SET status = ? WHERE goal_id = ?', ("–í –ø—Ä–æ—Ü–µ—Å—Å–µ", goal_id))
        await update.message.reply_text("–¶–µ–ª—å –æ—Å—Ç–∞–µ—Ç—Å—è –≤ –ø—Ä–æ—Ü–µ—Å—Å–µ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è.", reply_markup=get_goals_keyboard())
    conn.commit()
    conn.close()

    return ConversationHandler.END


import schedule
import time
from threading import Thread

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ—Ç–ø—Ä–∞–≤–∫–∏ –æ—Ç—á–µ—Ç–æ–≤
def send_reports():
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    cursor.execute('SELECT user_id FROM users')
    users = cursor.fetchall()
    conn.close()

    for user in users:
        user_id = user[0]
        conn = sqlite3.connect('users.db')
        cursor = conn.cursor()
        cursor.execute('SELECT description, progress FROM goals WHERE user_id = ?', (user_id,))
        goals = cursor.fetchall()
        conn.close()

        if goals:
            report_message = "üìä –í–∞—à –µ–∂–µ–Ω–µ–¥–µ–ª—å–Ω—ã–π –æ—Ç—á–µ—Ç –ø–æ —Ü–µ–ª—è–º:\n\n"
            for goal in goals:
                report_message += f"‚Ä¢ {goal[0]}\n–ü—Ä–æ–≥—Ä–µ—Å—Å: {goal[1]}%\n\n"

            # –û—Ç–ø—Ä–∞–≤–ª—è–µ–º –æ—Ç—á–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é
            context.bot.send_message(chat_id=user_id, text=report_message)

# –ó–∞–ø—É—Å–∫ –ø–ª–∞–Ω–∏—Ä–æ–≤—â–∏–∫–∞ –¥–ª—è –æ—Ç—á–µ—Ç–æ–≤
def start_scheduler():
    schedule.every().monday.at("09:00").do(send_reports)  # –ï–∂–µ–Ω–µ–¥–µ–ª—å–Ω–æ –≤ –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫ –≤ 9:00
    while True:
        schedule.run_pending()
        time.sleep(1)

# –ó–∞–ø—É—Å–∫ –ø–ª–∞–Ω–∏—Ä–æ–≤—â–∏–∫–∞ –≤ –æ—Ç–¥–µ–ª—å–Ω–æ–º –ø–æ—Ç–æ–∫–µ
scheduler_thread = Thread(target=start_scheduler, daemon=True)
scheduler_thread.start()




def main():
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö
    init_db()

    # –ó–∞–ø—É—Å–∫ –æ–±—Ä–∞–±–æ—Ç—á–∏–∫–∞ –æ—á–µ—Ä–µ–¥–∏ —Å–æ–æ–±—â–µ–Ω–∏–π –≤ –æ—Ç–¥–µ–ª—å–Ω–æ–º –ø–æ—Ç–æ–∫–µ
    queue_thread = threading.Thread(target=message_queue_worker, daemon=True)
    queue_thread.start()

    application = Application.builder().token(TOKEN).build()

    # Set bot commands in menu
    commands = [
        ("start", "–ü–µ—Ä–µ–∑–∞–ø—É—Å—Ç–∏—Ç—å –±–æ—Ç–∞")
    ]

    # Set commands when bot starts
    async def post_init(application: Application):
        await application.bot.set_my_commands(commands)

    application.post_init = post_init

    # –†–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏—è ConversationHandler –¥–ª—è —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
    conv_handler = ConversationHandler(
        entry_points=[CommandHandler("start", start)],
        states={
            WAITING_FOR_NAME: [MessageHandler(filters.TEXT & ~filters.COMMAND, handle_name)],
            WAITING_FOR_AGE: [MessageHandler(filters.TEXT & ~filters.COMMAND, handle_age)],
            WAITING_FOR_INTERESTS: [MessageHandler(filters.TEXT & ~filters.COMMAND, handle_interests)],
            WAITING_FOR_PROBLEMS: [MessageHandler(filters.TEXT & ~filters.COMMAND, handle_problems)],
        },
        fallbacks=[CommandHandler("cancel", cancel)],
    )

    # –†–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏—è ConversationHandler –¥–ª—è —É–ø—Ä–∞–≤–ª–µ–Ω–∏—è —Ü–µ–ª—è–º–∏
    # –†–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏—è ConversationHandler –¥–ª—è —É–ø—Ä–∞–≤–ª–µ–Ω–∏—è —Ü–µ–ª—è–º–∏
    goals_handler = ConversationHandler(
        entry_points=[CommandHandler("goals", goals)],
        states={
            "WAITING_FOR_GOAL_DESCRIPTION": [MessageHandler(filters.TEXT & ~filters.COMMAND, handle_goal_description)],
            "WAITING_FOR_GOAL_DEADLINE": [MessageHandler(filters.TEXT & ~filters.COMMAND, handle_goal_deadline)],
            "WAITING_FOR_COMPLETION_STATUS": [MessageHandler(filters.TEXT & ~filters.COMMAND, handle_completion_status)],
        },
        fallbacks=[CommandHandler("cancel", cancel)],
    )
    application.add_handler(goals_handler)

# –†–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏—è –æ–±—Ä–∞–±–æ—Ç—á–∏–∫–∞ –¥–ª—è –≤—ã–±–æ—Ä–∞ —Ü–µ–ª–∏
    application.add_handler(CallbackQueryHandler(handle_goal_selection, pattern="^complete_"))


    # –†–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏—è –æ–±—Ä–∞–±–æ—Ç—á–∏–∫–æ–≤
    application.add_handler(conv_handler)
    application.add_handler(CommandHandler("goals", goals))
    application.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, handle_message))
    application.add_handler(MessageHandler(filters.VOICE, handle_voice))
    application.add_handler(CallbackQueryHandler(handle_test_selection, pattern="^test_"))
    application.add_handler(CallbackQueryHandler(handle_test_answer, pattern="^answer_"))

    # –ó–∞–ø—É—Å–∫ –±–æ—Ç–∞
    application.run_polling()

if __name__ == '__main__':
    main()
```
